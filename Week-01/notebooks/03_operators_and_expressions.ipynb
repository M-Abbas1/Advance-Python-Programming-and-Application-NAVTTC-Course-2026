# Operators and Expressions

## Introduction

- Operators are special symbols in Python that perform operations on variables and values. 
- An expression is a combination of operators and operands (values or variables) that Python evaluates to produce a result. 
- Understanding operators is crucial for writing effective Python code, as they allow you to perform calculations, make comparisons, combine conditions, and manipulate data.

Think of operators as the verbs of programming - they describe actions to be performed on your data!

**What you'll learn:**
1. **Arithmetic Operators** - Mathematical operations (+, -, *, /, //, %, **)
2. **Relational (Comparison) Operators** - Comparing values (==, !=, >, <, >=, <=)
3. **Logical (Boolean) Operators** - Combining conditions (and, or, not)
4. **Membership Operators** - Testing membership (in, not in)
5. **Identity Operators** - Comparing object identity (is, is not)
6. **Bitwise Operators** - Binary operations (&, |, ^, ~, <<, >>)
7. **Assignment Operators** - Assigning values (=, +=, -=, *=, etc.)

Let's explore each type in detail!

## Part 1: Comprehensive Practice


## 1. Arithmetic Operators

Arithmetic operators are used to perform mathematical operations.

### 1.1 Basic Arithmetic Operators

````python
# Addition (+)
a = 10
b = 3
result = a + b
print(f"{a} + {b} = {result}")

# Subtraction (-)
result = a - b
print(f"{a} - {b} = {result}")

# Multiplication (*)
result = a * b
print(f"{a} * {b} = {result}")

# Division (/) - always returns float
result = a / b
print(f"{a} / {b} = {result}")
print(f"Type: {type(result)}")

# Even if numbers divide evenly
result = 10 / 2
print(f"10 / 2 = {result} (type: {type(result)})")
````

### 1.2 Floor Division and Modulus

````python
# Floor Division (//) - returns integer (quotient without remainder)
a = 10
b = 3
result = a // b
print(f"{a} // {b} = {result}")

# More examples
print(f"17 // 5 = {17 // 5}")   # 3
print(f"20 // 4 = {20 // 4}")   # 5
print(f"-10 // 3 = {-10 // 3}") # -4 (rounds down)

print()

# Modulus (%) - returns remainder
result = a % b
print(f"{a} % {b} = {result}")


# More examples
print(f"17 % 5 = {17 % 5}")   # 2
print(f"20 % 4 = {20 % 4}")   # 0
print(f"100 % 7 = {100 % 7}") # 2
````

### 1.3 Exponentiation

````python
# Exponentiation (**) - power operation
base = 2
exponent = 3
result = base ** exponent
print(f"{base} ** {exponent} = {result}")

# More examples
print(f"5 ** 2 = {5 ** 2}")     # 25
print(f"10 ** 3 = {10 ** 3}")   # 1000
print(f"2 ** 8 = {2 ** 8}")     # 256

# Square root using exponent
number = 16
square_root = number ** 0.5
print(f"Square root of {number} = {square_root}")

# Cube root
number = 27
cube_root = number ** (1/3)
print(f"Cube root of {number} = {cube_root:.2f}")
````

### 1.4 Operator Precedence

````python
# Order of operations: PEMDAS
# Parentheses, Exponents, Multiplication/Division, Addition/Subtraction

# Without parentheses
result = 2 + 3 * 4
print(f"2 + 3 * 4 = {result}")  # 14 (multiplication first)

# With parentheses
result = (2 + 3) * 4
print(f"(2 + 3) * 4 = {result}")  # 20 (addition first)

# Complex expression
result = 2 ** 3 + 4 * 5 - 6 / 2
print(f"2 ** 3 + 4 * 5 - 6 / 2 = {result}")  # 25.0

# Breaking it down:
print("\nStep by step:")
print(f"1. 2 ** 3 = {2 ** 3}")           # 8
print(f"2. 4 * 5 = {4 * 5}")             # 20
print(f"3. 6 / 2 = {6 / 2}")             # 3.0
print(f"4. 8 + 20 - 3.0 = {8 + 20 - 3.0}")  # 25.0
````

### 1.5 Unary Operators

````python
# Unary plus (+) - doesn't change the value
x = 5
result = +x
print(f"+{x} = {result}")

# Unary minus (-) - negates the value
x = 5
result = -x
print(f"-{x} = {result}")

# Double negative
x = -5
result = -x
print(f"-({x}) = {result}")
````

## 2. Relational (Comparison) Operators

Relational operators compare two values and return a boolean (True or False).

### 2.1 Basic Comparison Operators

````python
a = 10
b = 5

# Equal to (==)
print(f"{a} == {b}: {a == b}")  # False
print(f"{a} == 10: {a == 10}")  # True

# Not equal to (!=)
print(f"{a} != {b}: {a != b}")  # True
print(f"{a} != 10: {a != 10}")  # False

# Greater than (>)
print(f"{a} > {b}: {a > b}")    # True
print(f"{b} > {a}: {b > a}")    # False

# Less than (<)
print(f"{a} < {b}: {a < b}")    # False
print(f"{b} < {a}: {b < a}")    # True

# Greater than or equal to (>=)
print(f"{a} >= {b}: {a >= b}")  # True
print(f"{a} >= 10: {a >= 10}")  # True
print(f"{a} >= 15: {a >= 15}")  # False

# Less than or equal to (<=)
print(f"{b} <= {a}: {b <= a}")  # True
print(f"{a} <= 10: {a <= 10}")  # True
print(f"{a} <= 5: {a <= 5}")    # False
````

### 2.2 Comparing Different Data Types

````python
# Comparing integers and floats
print(f"10 == 10.0: {10 == 10.0}")  # True (values are equal)
print(f"5 < 5.5: {5 < 5.5}")        # True

# Comparing strings
name1 = "Alice"
name2 = "Bob"
print(f"'{name1}' == '{name2}': {name1 == name2}")  # False
print(f"'{name1}' < '{name2}': {name1 < name2}")    # True (alphabetical)

# Case sensitivity
print(f"'Alice' == 'alice': {'Alice' == 'alice'}")  # False
print(f"'Alice'.lower() == 'alice': {'Alice'.lower() == 'alice'}")  # True

# Comparing booleans
print(f"True == 1: {True == 1}")    # True
print(f"False == 0: {False == 0}")  # True
print(f"True > False: {True > False}")  # True
````

### 2.3 Chaining Comparisons

````python
# Python allows chaining comparisons
x = 5

# Check if x is between 1 and 10
result = 1 < x < 10
print(f"1 < {x} < 10: {result}")  # True

# Equivalent to:
result = (1 < x) and (x < 10)
print(f"(1 < {x}) and ({x} < 10): {result}")  # True

# More examples
age = 25
print(f"18 <= {age} <= 65: {18 <= age <= 65}")  # True

score = 85
print(f"60 <= {score} < 70: {60 <= score < 70}")  # False
print(f"80 <= {score} < 90: {80 <= score < 90}")  # True
````

## 3. Logical (Boolean) Operators

Logical operators are used to combine conditional statements.

### 3.1 AND Operator

````python
# AND operator - both conditions must be True
print("=== AND Operator ===")
print(f"True and True: {True and True}")      # True
print(f"True and False: {True and False}")    # False
print(f"False and True: {False and True}")    # False
print(f"False and False: {False and False}")  # False

print()

# Practical example
age = 25
has_license = True

can_drive = (age >= 18) and has_license
print(f"Can drive? {can_drive}")  # True

# Another example
temperature = 75
is_sunny = True

good_weather = (temperature > 70) and is_sunny
print(f"Good weather? {good_weather}")  # True
````

### 3.2 OR Operator

````python
# OR operator - at least one condition must be True
print("=== OR Operator ===")
print(f"True or True: {True or True}")      # True
print(f"True or False: {True or False}")    # True
print(f"False or True: {False or True}")    # True
print(f"False or False: {False or False}")  # False

print()

# Practical example
is_weekend = True
is_holiday = False

can_relax = is_weekend or is_holiday
print(f"Can relax? {can_relax}")  # True

# Another example
has_coupon = False
is_member = True

gets_discount = has_coupon or is_member
print(f"Gets discount? {gets_discount}")  # True
````

### 3.3 NOT Operator

````python
# NOT operator - inverts the boolean value
print("=== NOT Operator ===")
print(f"not True: {not True}")    # False
print(f"not False: {not False}")  # True

print()

# Practical example
is_raining = False
is_sunny = not is_raining
print(f"Is sunny? {is_sunny}")  # True

# With comparisons
x = 5
print(f"not (x > 10): {not (x > 10)}")  # True
print(f"not (x == 5): {not (x == 5)}")  # False

# Double negative
value = True
print(f"not not {value}: {not not value}")  # True
````

### 3.4 Combining Logical Operators

````python
# Complex conditions
age = 25
has_license = True
has_car = False

# AND with OR
result = (age >= 18 and has_license) or has_car
print(f"Result 1: {result}")  # True

# Using parentheses for clarity
score = 85
attendance = 90

passes = (score >= 60) and (attendance >= 75)
print(f"Student passes? {passes}")  # True

# More complex example
temperature = 75
is_weekend = True
has_plans = False

go_outside = (temperature > 70 and is_weekend) and not has_plans
print(f"Go outside? {go_outside}")  # True
````

### 3.5 Short-Circuit Evaluation

````python
# Python uses short-circuit evaluation
# For AND: if first is False, doesn't evaluate second
# For OR: if first is True, doesn't evaluate second

# AND short-circuit
x = 0
result = (x != 0) and (10 / x > 5)  # Doesn't cause error
print(f"AND short-circuit: {result}")  # False

# If we didn't have short-circuit, this would cause division by zero error
# Uncomment to see error:
# result = (10 / x > 5) and (x != 0)  # Error!

# OR short-circuit
has_permission = True
# result = has_permission or (1/0 > 0)  # Would not evaluate second part
print(f"OR short-circuit example: {has_permission or False}")  # True
````

## 4. Membership Operators

Membership operators test whether a value is found in a sequence (string, list, tuple, etc.).

### 4.1 The 'in' Operator

````python
# Check if value is in a string
text = "Python Programming"
print(f"'Python' in '{text}': {'Python' in text}")  # True
print(f"'Java' in '{text}': {'Java' in text}")      # False
print(f"'python' in '{text}': {'python' in text}")  # False (case-sensitive)

print()

# Check if value is in a list
fruits = ["apple", "banana", "cherry"]
print(f"'apple' in {fruits}: {'apple' in fruits}")    # True
print(f"'orange' in {fruits}: {'orange' in fruits}")  # False

print()

# Check if value is in a tuple
numbers = (1, 2, 3, 4, 5)
print(f"3 in {numbers}: {3 in numbers}")  # True
print(f"6 in {numbers}: {6 in numbers}")  # False

print()

# Check if key is in a dictionary
person = {"name": "Alice", "age": 25}
print(f"'name' in {person}: {'name' in person}")      # True
print(f"'address' in {person}: {'address' in person}")  # False
````

### 4.2 The 'not in' Operator

````python
# Check if value is NOT in a sequence
text = "Python Programming"
print(f"'Java' not in '{text}': {'Java' not in text}")  # True
print(f"'Python' not in '{text}': {'Python' not in text}")  # False

print()

# With lists
colors = ["red", "blue", "green"]
print(f"'yellow' not in {colors}: {'yellow' not in colors}")  # True
print(f"'red' not in {colors}: {'red' not in colors}")        # False

print()

# Practical example
allowed_users = ["alice", "bob", "charlie"]
username = "dave"

if username not in allowed_users:
    print(f"{username} is not an allowed user")
else:
    print(f"{username} is an allowed user")
````

## 5. Identity Operators

Identity operators compare the memory locations of two objects (not their values).

### 5.1 The 'is' Operator

````python
# 'is' checks if two variables point to the same object in memory
a = [1, 2, 3]
b = a  # b points to the same list as a
c = [1, 2, 3]  # c is a new list with same values

print(f"a is b: {a is b}")  # True (same object)
print(f"a is c: {a is c}")  # False (different objects)
print(f"a == c: {a == c}")  # True (same values)

print()

# With None
x = None
y = None
print(f"x is None: {x is None}")  # True
print(f"y is None: {y is None}")  # True
print(f"x is y: {x is y}")        # True (None is a singleton)

print()

# With small integers (Python caches small integers)
x = 5
y = 5
print(f"x is y: {x is y}")  # True (cached)

# With large integers
x = 1000
y = 1000
print(f"x is y (large numbers): {x is y}")  # May be False (not cached)
````

### 5.2 The 'is not' Operator

````python
# 'is not' checks if two variables point to different objects
a = [1, 2, 3]
b = [1, 2, 3]

print(f"a is not b: {a is not b}")  # True (different objects)

# With None
x = 10
print(f"x is not None: {x is not None}")  # True

# Common use case: checking for None
value = None
if value is None:
    print("Value is None")

value = 42
if value is not None:
    print("Value is not None")
````

### 5.3 Understanding id() Function

````python
# id() returns the memory address of an object
a = [1, 2, 3]
b = a
c = [1, 2, 3]

print(f"id(a): {id(a)}")
print(f"id(b): {id(b)}")
print(f"id(c): {id(c)}")

print()
print(f"a is b: {a is b}  # Same id? {id(a) == id(b)}")
print(f"a is c: {a is c}  # Same id? {id(a) == id(c)}")
````

## 6. Bitwise Operators

Bitwise operators work on binary representations of numbers.

### 6.1 Bitwise AND (&)

````python
# Bitwise AND: 1 if both bits are 1, else 0
a = 12  # Binary: 1100
b = 10  # Binary: 1010
result = a & b  # Binary: 1000 = 8

print(f"{a} & {b} = {result}")
print(f"Binary: {bin(a)} & {bin(b)} = {bin(result)}")

# More examples
print(f"5 & 3 = {5 & 3}")   # 0101 & 0011 = 0001 = 1
print(f"7 & 3 = {7 & 3}")   # 0111 & 0011 = 0011 = 3
````

### 6.2 Bitwise OR (|)

````python
# Bitwise OR: 1 if at least one bit is 1
a = 12  # Binary: 1100
b = 10  # Binary: 1010
result = a | b  # Binary: 1110 = 14

print(f"{a} | {b} = {result}")
print(f"Binary: {bin(a)} | {bin(b)} = {bin(result)}")

# More examples
print(f"5 | 3 = {5 | 3}")   # 0101 | 0011 = 0111 = 7
print(f"8 | 4 = {8 | 4}")   # 1000 | 0100 = 1100 = 12
````

### 6.3 Bitwise XOR (^)

````python
# Bitwise XOR: 1 if bits are different
a = 12  # Binary: 1100
b = 10  # Binary: 1010
result = a ^ b  # Binary: 0110 = 6

print(f"{a} ^ {b} = {result}")
print(f"Binary: {bin(a)} ^ {bin(b)} = {bin(result)}")

# More examples
print(f"5 ^ 3 = {5 ^ 3}")   # 0101 ^ 0011 = 0110 = 6
print(f"7 ^ 7 = {7 ^ 7}")   # 0111 ^ 0111 = 0000 = 0

# XOR trick: swapping without temp variable
x = 5
y = 10
print(f"Before swap: x = {x}, y = {y}")
x = x ^ y
y = x ^ y
x = x ^ y
print(f"After swap: x = {x}, y = {y}")
````

### 6.4 Bitwise NOT (~)

````python
# Bitwise NOT: inverts all bits (returns -(n+1))
a = 5   # Binary: 0101
result = ~a  # -6

print(f"~{a} = {result}")

# More examples
print(f"~10 = {~10}")  # -11
print(f"~0 = {~0}")    # -1
print(f"~(-1) = {~(-1)}")  # 0
````

### 6.5 Left Shift (<<) and Right Shift (>>)

````python
# Left shift: shifts bits to the left (multiply by 2^n)
a = 5  # Binary: 0101
result = a << 1  # Binary: 1010 = 10
print(f"{a} << 1 = {result}")  # 5 * 2 = 10

result = a << 2  # Binary: 10100 = 20
print(f"{a} << 2 = {result}")  # 5 * 4 = 20

print()

# Right shift: shifts bits to the right (divide by 2^n)
a = 20  # Binary: 10100
result = a >> 1  # Binary: 1010 = 10
print(f"{a} >> 1 = {result}")  # 20 / 2 = 10

result = a >> 2  # Binary: 101 = 5
print(f"{a} >> 2 = {result}")  # 20 / 4 = 5

# Quick multiplication and division by powers of 2
print(f"16 << 2 = {16 << 2}")  # 16 * 4 = 64
print(f"100 >> 3 = {100 >> 3}")  # 100 / 8 = 12
````

## 7. Assignment Operators

Assignment operators are used to assign values to variables.

### 7.1 Basic Assignment (=)

````python
# Basic assignment
x = 10
print(f"x = {x}")

# Assigning from expression
y = x + 5
print(f"y = {y}")

# Multiple assignment
a = b = c = 0
print(f"a = {a}, b = {b}, c = {c}")
````

### 7.2 Compound Assignment Operators

````python
# Add and assign (+=)
x = 10
x += 5  # Same as: x = x + 5
print(f"After x += 5: {x}")

# Subtract and assign (-=)
x = 10
x -= 3  # Same as: x = x - 3
print(f"After x -= 3: {x}")

# Multiply and assign (*=)
x = 10
x *= 2  # Same as: x = x * 2
print(f"After x *= 2: {x}")

# Divide and assign (/=)
x = 10
x /= 2  # Same as: x = x / 2
print(f"After x /= 2: {x}")

# Floor divide and assign (//=)
x = 10
x //= 3  # Same as: x = x // 3
print(f"After x //= 3: {x}")

# Modulus and assign (%=)
x = 10
x %= 3  # Same as: x = x % 3
print(f"After x %= 3: {x}")

# Exponent and assign (**=)
x = 2
x **= 3  # Same as: x = x ** 3
print(f"After x **= 3: {x}")
````

### 7.3 Bitwise Compound Assignment

````python
# Bitwise AND and assign (&=)
x = 12  # 1100
x &= 10  # Same as: x = x & 10
print(f"After x &= 10: {x}")

# Bitwise OR and assign (|=)
x = 12  # 1100
x |= 10  # Same as: x = x | 10
print(f"After x |= 10: {x}")

# Bitwise XOR and assign (^=)
x = 12  # 1100
x ^= 10  # Same as: x = x ^ 10
print(f"After x ^= 10: {x}")

# Left shift and assign (<<=)
x = 5
x <<= 2  # Same as: x = x << 2
print(f"After x <<= 2: {x}")

# Right shift and assign (>>=)
x = 20
x >>= 2  # Same as: x = x >> 2
print(f"After x >>= 2: {x}")
````

### 7.4 Practical Examples with Assignment Operators

````python
# Counter example
count = 0
count += 1  # Increment
count += 1
count += 1
print(f"Count: {count}")

# Score accumulation
total_score = 0
total_score += 50  # Level 1
total_score += 75  # Level 2
total_score += 100  # Level 3
print(f"Total Score: {total_score}")

# String concatenation
message = "Hello"
message += " "
message += "World"
print(f"Message: {message}")

# List extension
numbers = [1, 2, 3]
numbers += [4, 5]
print(f"Numbers: {numbers}")
````

### 7.5 Walrus Operator (:=) - Python 3.8+

````python
# Walrus operator: assigns and returns value in same expression

# Without walrus operator
name = "Alice"
if len(name) > 3:
    print(f"{name} has more than 3 characters")

# With walrus operator
if (n := len("Alice")) > 3:
    print(f"Length {n} is more than 3")

# Useful in loops (conceptual example)
data = [1, 2, 3, 4, 5]
if (length := len(data)) > 3:
    print(f"Data has {length} elements")
````
