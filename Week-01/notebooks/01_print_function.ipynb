## Part 1: Comprehensive Practice


### 1.1 Basic Print Statement

````python
# Simple print statement
print("Hello, World!")

# Print numbers
print(42)
print(3.14159)
````

### 1.2 Printing Multiple Values

````python
# Print multiple items separated by comma
print("Python", "is", "awesome")

# Print mix of strings and numbers
print("I am", 25, "years old")

# Print multiple data types
print("Integer:", 100, "Float:", 3.14, "String:", "Hello")
````

### 1.3 The 'sep' Parameter (Separator)

````python
# Default separator is a space
print("Apple", "Banana", "Cherry")

# Custom separator with dash
print("Apple", "Banana", "Cherry", sep="-")

# Custom separator with comma
print("Apple", "Banana", "Cherry", sep=", ")

# No separator
print("Apple", "Banana", "Cherry", sep="")

# Separator with newline
print("First", "Second", "Third", sep="\n")

# Separator with multiple characters
print("Python", "Java", "C++", sep=" | ")
````

### 1.4 The 'end' Parameter

````python
# Default end is newline (\n)
print("Hello")
print("World")

print("---")  

# Custom end parameter - no newline
print("Hello", end=" ")
print("World")

print("---")  

# End with different characters
print("Python", end="***")
print("Programming")

print("---")  

# Combining multiple prints on same line
print("Loading", end="")
print(".", end="")
print(".", end="")
print(".")
````

### 1.5 Combining 'sep' and 'end' Parameters

````python
# Using both sep and end together
print("Apple", "Banana", "Cherry", sep=" | ", end=" <-- Fruits\n")

# Creating formatted output
print("Name", "Age", "City", sep=" - ", end=" (Header)\n")
print("John", "25", "New York", sep=" - ", end=" (Data)\n")
````

### 1.6 Escape Sequences

````python
# Newline (\n)
print("First Line\nSecond Line\nThird Line")

print("---")  

# Tab (\t)
print("Name:\tJohn")
print("Age:\t25")
print("City:\tNew York")

print("---")  

# Backslash (\\)
print("This is a backslash: \\")

# Single quote (\')
print('It\'s a beautiful day')

# Double quote (\")
print("She said, \"Hello!\"")

print("---")  # Separator

# Carriage return (\r) - moves cursor to beginning
print("Hello\rWorld")  # World overwrites Hello

# Backspace (\b)
print("Helloo\b")
````

### 1.7 Raw Strings

````python
# Regular string with escape sequences
print("C:\new\test")

# Raw string (r prefix) - ignores escape sequences
print(r"C:\new\test")

# Useful for file paths 
print(r"C:\Users\Student\Documents\file.txt")
````

### 1.8 Printing Different Data Types

````python
# Integer
print(42)

# Float
print(3.14159)

# String
print("Hello, Python!")

# Boolean
print(True)
print(False)

# List
print([1, 2, 3, 4, 5])

# Tuple
print((10, 20, 30))

# Dictionary
print({"name": "John", "age": 25})

# None
print(None)
````

### 1.9 String Concatenation in Print

````python
# Using + operator (only works with strings)
print("Hello" + " " + "World")

# This will cause error - cannot concatenate string and number
# print("I am " + 25 + " years old")  # Uncomment to see error

# Solution: Convert number to string
print("I am " + str(25) + " years old")

# Better approach: use comma (automatic conversion)
print("I am", 25, "years old")
````

### 1.10 String Formatting Methods

````python
# Method 1: Using % operator (old style)
name = "Alice"
age = 30
print("My name is %s and I am %d years old" % (name, age))

# Method 2: Using .format() method
print("My name is {} and I am {} years old".format(name, age))
print("My name is {0} and I am {1} years old. Yes, {0}!".format(name, age))
print("My name is {n} and I am {a} years old".format(n=name, a=age))

# Method 3: Using f-strings (modern and recommended)
print(f"My name is {name} and I am {age} years old")

# F-strings with expressions
print(f"Next year I will be {age + 1} years old")
print(f"My name in uppercase: {name.upper()}")
````

### 1.11 Formatting Numbers in Print

````python
# Floating point precision
pi = 3.14159265359

print(f"Pi with 2 decimals: {pi:.2f}")
print(f"Pi with 4 decimals: {pi:.4f}")

# Large numbers with comma separator
big_number = 1234567890
print(f"Formatted number: {big_number:,}")

# Percentage formatting
score = 0.856
print(f"Score: {score:.1%}")

# Scientific notation
print(f"Scientific: {big_number:.2e}")
````

### 1.12 Print with Width and Alignment

````python
# Right alignment (default for numbers)
print(f"{'Python':>15}")
print(f"{42:>15}")

# Left alignment
print(f"{'Python':<15}***")
print(f"{42:<15}***")

# Center alignment
print(f"{'Python':^15}")
print(f"{42:^15}")

# With fill characters
print(f"{'Python':*^20}")
print(f"{'Hello':-<20}")
print(f"{100:0>10}")
````

### 1.13 Printing Empty Lines and Special Cases

````python
# Print empty line
print()

# Print multiple empty lines
print("\n\n\n")

# Print nothing (empty string)
print("")

# Print just spaces
print("     ")
````

### 1.14 Practical Examples

````python
# Creating a simple receipt
print("="*40)
print("RECEIPT".center(40))
print("="*40)
print(f"{'Item':<20} {'Price':>10}")
print("-"*40)
print(f"{'Apple':<20} ${3.99:>9.2f}")
print(f"{'Banana':<20} ${2.50:>9.2f}")
print(f"{'Orange':<20} ${4.25:>9.2f}")
print("-"*40)
print(f"{'Total':<20} ${10.74:>9.2f}")
print("="*40)

print()  # Empty line

# Creating a simple table
print(f"{'Name':<15} {'Age':>5} {'City':<15}")
print("-" * 37)
print(f"{'Alice':<15} {25:>5} {'New York':<15}")
print(f"{'Bob':<15} {30:>5} {'Los Angeles':<15}")
print(f"{'Charlie':<15} {28:>5} {'Chicago':<15}")
````
