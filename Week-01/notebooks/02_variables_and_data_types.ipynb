### 1.1 Creating Variables

````python
# Creating a variable - basic syntax: variable_name = value
age = 25
print(age)

# Creating multiple variables
name = "Alice"
height = 5.6
is_student = True

print(name)
print(height)
print(is_student)
````

### 1.2 Variable Naming Rules and Conventions

````python
# Variables Naming Conventions
# snamke_case
# camelCase
# PascalCase

# Valid variable names
user_name = "John"
user_age = 30
user123 = "User with numbers"
_private_var = "Underscore prefix"
userName = "Camel case (not recommended in Python)"

print(user_name, user_age, user123, _private_var, userName)

# Invalid variable names (uncomment to see errors)
# 123user = "Cannot start with number"
# user-name = "Cannot use hyphens"
# user name = "Cannot have spaces"
# for = "Cannot use reserved keywords"

# Python naming convention: snake_case (recommended)
first_name = "Alice"
last_name = "Smith"
total_amount = 150.75
is_active = True

print(f"Best practice: {first_name} {last_name}")
````

### 1.3 Variable Assignment and Reassignment

````python
# Initial assignment
x = 10
print("Initial value:", x)

# Reassigning a new value
x = 20
print("After reassignment:", x)

# Variables can change types (dynamic typing)
x = "Now I'm a string!"
print("After type change:", x)

# Using a variable in calculation
score = 85
score = score + 10  # Add 10 to current score
print("New score:", score)
````

### 1.4 Data Type: Integer (int)

````python
# Integers are whole numbers (positive, negative, or zero)
positive_num = 42
negative_num = -17
zero = 0

print("Positive:", positive_num)
print("Negative:", negative_num)
print("Zero:", zero)

# Large integers (Python handles any size)
big_number = 999999999999999999999999999
print("Big number:", big_number)

# Integer operations
a = 10
b = 3

print(f"Addition: {a} + {b} = {a + b}")
print(f"Subtraction: {a} - {b} = {a - b}")
print(f"Multiplication: {a} * {b} = {a * b}")
print(f"Division: {a} / {b} = {a / b}")  # Returns float
print(f"Floor Division: {a} // {b} = {a // b}")  # Returns integer
print(f"Modulus (remainder): {a} % {b} = {a % b}")
print(f"Exponentiation: {a} ** {b} = {a ** b}")
````

### 1.5 Data Type: Float (float)

````python
# Floats are decimal numbers
temperature = 98.6
price = 19.99
pi = 3.14159
negative_float = -45.7

print("Temperature:", temperature)
print("Price:", price)
print("Pi:", pi)
print("Negative float:", negative_float)

# Scientific notation
small_number = 1.5e-4  # 1.5 × 10^-4 = 0.00015
large_number = 3.5e6   # 3.5 × 10^6 = 3500000

print("Small number:", small_number)
print("Large number:", large_number)

# Float operations
x = 10.5
y = 2.5

print(f"Addition: {x} + {y} = {x + y}")
print(f"Subtraction: {x} - {y} = {x - y}")
print(f"Multiplication: {x} * {y} = {x * y}")
print(f"Division: {x} / {y} = {x / y}")

# Mixing integers and floats
result = 10 + 5.5  # Result will be float
print(f"Integer + Float = {result} (type: float)")
````

### 1.6 Data Type: String (str)

````python
# Strings are sequences of characters
# Can use single or double quotes
message1 = "Hello, World!"
message2 = 'Hello, Python!'

print(message1)
print(message2)

# String with quotes inside
quote1 = "She said, 'Hello!'"
quote2 = 'He said, "Goodbye!"'

print(quote1)
print(quote2)

# Multi-line strings (triple quotes)
long_text = """This is a
multi-line
string."""

print(long_text)

# Empty string
empty = ""
print(f"Empty string: '{empty}'")

# String with numbers (still a string!)
zip_code = "90210"
phone = "555-1234"
print(f"Zip code: {zip_code}, Phone: {phone}")
````

### 1.7 String Operations

````python
# String concatenation (joining)
first_name = "John"
last_name = "Doe"
full_name = first_name + " " + last_name
print("Full name:", full_name)

# String repetition
laugh = "Ha" * 5
print(laugh)

line = "=" * 30
print(line)

# String length
text = "Python Programming"
print(f"The length of '{text}' is {len(text)} characters")

# String indexing (accessing individual characters)
word = "Python"
print(f"First character: {word[0]}")
print(f"Second character: {word[1]}")
print(f"Last character: {word[-1]}")
print(f"Second to last: {word[-2]}")

# String slicing
text = "Hello, World!"
print(f"First 5 characters: {text[0:5]}")
print(f"Characters from index 7: {text[7:]}")
print(f"Characters up to index 5: {text[:5]}")
print(f"Last 6 characters: {text[-6:]}")
````

### 1.8 String Methods

````python
# Common string methods
text = "  Hello, Python!  "

# Case conversion
print(f"Upper case: {text.upper()}")
print(f"Lower case: {text.lower()}")
print(f"Title case: {text.title()}")
print(f"Capitalize: {text.capitalize()}")

# Removing whitespace
print(f"Strip whitespace: '{text.strip()}'")
print(f"Left strip: '{text.lstrip()}'")
print(f"Right strip: '{text.rstrip()}'")

# Replace
message = "Hello, World!"
new_message = message.replace("World", "Python")
print(f"Original: {message}")
print(f"Replaced: {new_message}")

# Split (convert string to list)
sentence = "Python is awesome"
words = sentence.split()
print(f"Split: {words}")

# Count occurrences
text = "banana"
print(f"Number of 'a' in 'banana': {text.count('a')}")

# Find position
sentence = "Python is fun"
position = sentence.find("is")
print(f"Position of 'is': {position}")
````

### 1.9 Data Type: Boolean (bool)

````python
# Booleans represent True or False values
is_sunny = True
is_raining = False

print("Is it sunny?", is_sunny)
print("Is it raining?", is_raining)

# Booleans from comparisons
age = 18
is_adult = age >= 18
print(f"Is adult? {is_adult}")

# Comparison operators
x = 10
y = 5

print(f"{x} == {y}: {x == y}")  # Equal to
print(f"{x} != {y}: {x != y}")  # Not equal to
print(f"{x} > {y}: {x > y}")    # Greater than
print(f"{x} < {y}: {x < y}")    # Less than
print(f"{x} >= {y}: {x >= y}")  # Greater than or equal to
print(f"{x} <= {y}: {x <= y}")  # Less than or equal to

# Boolean with strings
name1 = "Alice"
name2 = "Bob"
print(f"'{name1}' == '{name2}': {name1 == name2}")
print(f"'{name1}' != '{name2}': {name1 != name2}")
````

### 1.10 Logical Operators with Booleans

````python
# Logical operators: and, or, not

# AND operator (both must be True)
has_ticket = True
has_id = True
can_enter = has_ticket and has_id
print(f"Can enter? {can_enter}")

has_ticket = True
has_id = False
can_enter = has_ticket and has_id
print(f"Can enter? {can_enter}")

# OR operator (at least one must be True)
is_weekend = True
is_holiday = False
can_relax = is_weekend or is_holiday
print(f"Can relax? {can_relax}")

# NOT operator (inverts the boolean)
is_raining = False
is_sunny = not is_raining
print(f"Is sunny? {is_sunny}")

# Combining logical operators
age = 25
has_license = True
can_rent_car = (age >= 21) and has_license
print(f"Can rent car? {can_rent_car}")
````

### 1.11 Checking Data Types - type() Function

````python
# Using type() to check data types
age = 25
price = 19.99
name = "Alice"
is_student = True

print(f"Type of {age}: {type(age)}")
print(f"Type of {price}: {type(price)}")
print(f"Type of '{name}': {type(name)}")
print(f"Type of {is_student}: {type(is_student)}")

# Checking types in conditions
value = 42
if type(value) == int:
    print(f"{value} is an integer")

# Alternative: isinstance()
print(f"Is 42 an integer? {isinstance(42, int)}")
print(f"Is 'Hello' a string? {isinstance('Hello', str)}")
print(f"Is 3.14 a float? {isinstance(3.14, float)}")
````

### 1.12 Type Conversion (Casting)

````python
# Converting between types

# Integer to Float
num_int = 10
num_float = float(num_int)
print(f"Integer {num_int} to Float: {num_float}")

# Float to Integer (truncates decimal part)
num_float = 10.7
num_int = int(num_float)
print(f"Float {num_float} to Integer: {num_int}")

# String to Integer
str_num = "100"
int_num = int(str_num)
print(f"String '{str_num}' to Integer: {int_num}")
print(f"Type: {type(int_num)}")

# String to Float
str_num = "3.14"
float_num = float(str_num)
print(f"String '{str_num}' to Float: {float_num}")

# Integer/Float to String
age = 25
age_str = str(age)
print(f"Integer {age} to String: '{age_str}'")
print(f"Type: {type(age_str)}")

# Boolean to Integer
bool_true = True
bool_false = False
print(f"True to Integer: {int(bool_true)}")   # 1
print(f"False to Integer: {int(bool_false)}") # 0

# Integer to Boolean
print(f"0 to Boolean: {bool(0)}")      # False
print(f"1 to Boolean: {bool(1)}")      # True
print(f"42 to Boolean: {bool(42)}")    # True
print(f"-5 to Boolean: {bool(-5)}")    # True

# String to Boolean
print(f"Empty string to Boolean: {bool('')}")      # False
print(f"Non-empty string to Boolean: {bool('Hi')}") # True
````

### 1.13 Type Conversion Errors

````python
# Some conversions will cause errors

# This works
num_str = "123"
num = int(num_str)
print(f"Successful conversion: {num}")

# This will cause an error (uncomment to see)
# invalid_str = "Hello"
# num = int(invalid_str)  # ValueError: invalid literal for int()

# This will also cause an error (uncomment to see)
# decimal_str = "3.14"
# num = int(decimal_str)  # ValueError: invalid literal for int()

# Solution: convert to float first, then to int
decimal_str = "3.14"
num = int(float(decimal_str))
print(f"Converted '{decimal_str}' to integer: {num}")
````

### 1.14 Multiple Assignment

````python
# Assign same value to multiple variables
x = y = z = 0
print(f"x = {x}, y = {y}, z = {z}")

# Assign multiple values at once
name, age, city = "Alice", 25, "New York"
print(f"Name: {name}, Age: {age}, City: {city}")

# Unpacking
a, b, c = 1, 2, 3
print(f"a = {a}, b = {b}, c = {c}")

# Swapping variables
x = 10
y = 20
print(f"Before swap: x = {x}, y = {y}")

x, y = y, x  # Python magic!
print(f"After swap: x = {x}, y = {y}")
````

### 1.15 Constants (Convention)

````python
# Python doesn't have true constants, but we use UPPER_CASE by convention
# to indicate a value should not be changed

PI = 3.14159
GRAVITY = 9.8
MAX_SPEED = 120
COMPANY_NAME = "Tech Corp"

print(f"Pi: {PI}")
print(f"Gravity: {GRAVITY}")
print(f"Max Speed: {MAX_SPEED}")
print(f"Company: {COMPANY_NAME}")

# These can technically be changed, but shouldn't be
# PI = 3.14  # Don't do this!
````

### 1.17 Practical Examples
