## 1. 🤔 What is Object-Oriented Programming?

So far in **Structured Programming**, we wrote code as a sequence of instructions — functions that take inputs and return outputs. That works great for many problems.

But as programs grow larger, we need a better way to **organize** and **model** the real world.

**Object-Oriented Programming (OOP)** is a programming paradigm that organizes code around **objects** — bundles of **data (attributes)** and **behavior (methods)** that belong together.

### Real-World Analogy

Think of a **bank account**:
- It has **data**: account number, owner name, balance
- It has **behaviors**: deposit, withdraw, check balance

Instead of keeping these as separate variables and functions, OOP lets us package them together.

### The 4 Pillars of OOP

```
┌─────────────────────────────────────────────────────┐
│                  4 Pillars of OOP                   │
│                                                     │
│  1. Encapsulation  — Bundling data + behavior       │
│  2. Inheritance    — Reusing & extending classes    │
│  3. Polymorphism   — One interface, many forms      │
│  4. Abstraction    — Hiding complexity              │
└─────────────────────────────────────────────────────┘
```

> In this notebook we focus on **Encapsulation** and the foundations. Inheritance and Polymorphism are covered in the next notebook.

## 2. 🏗️ Classes and Objects

### Class vs Object

| Term | Analogy | In Python |
|---|---|---|
| **Class** | Blueprint / Cookie cutter | The definition written with `class` |
| **Object** | The actual house / cookie | An **instance** created from the class |

One class → many objects. Each object is independent.

```
  Class: Dog
  ┌─────────────┐
  │  Blueprint  │──────► dog1 = Dog("Rex",  3)
  │  name       │──────► dog2 = Dog("Bella", 5)
  │  age        │──────► dog3 = Dog("Max",  1)
  │  bark()     │
  └─────────────┘
```

### Defining a Class

````python
# Simplest possible class
class Dog:
    pass  # 'pass' means the class body is empty for now

# Creating objects (instances)
dog1 = Dog()
dog2 = Dog()

print(type(dog1))       # <class '__main__.Dog'>
print(dog1)             # default representation
print(dog1 is dog2)     # False — they are different objects
````

## 3. 🚀 The `__init__` Method (Constructor)

The `__init__` method is called **automatically** when you create a new object. It is used to **initialize** (set up) the object's data.

- Methods that start and end with double underscores `__` are called **dunder (magic) methods**.
- `__init__` stands for *initialize*.

````python
class Dog:
    def __init__(self, name, age, breed):
        # self.attribute = value  ← this is how you store data on an object
        self.name  = name
        self.age   = age
        self.breed = breed

# Creating instances — __init__ is called automatically
dog1 = Dog("Rex", 3, "Labrador")
dog2 = Dog("Bella", 5, "Poodle")

# Accessing attributes with dot notation
print(dog1.name, dog1.age, dog1.breed)
print(dog2.name, dog2.age, dog2.breed)
````

````python
# ✅ Each object holds its OWN data independently
dog1.age = 4          # changing dog1's age
print(dog1.age)       # 4
print(dog2.age)       # 5 — dog2 is NOT affected
````

## 4. 🔧 `self` — What Is It?

`self` is a **reference to the current object**. Python passes it automatically when you call a method on an object.

```python
dog1.bark()        # Python translates this to:
Dog.bark(dog1)     # same thing! 'self' = dog1
```

- `self` must be the **first parameter** of every instance method.
- The name `self` is a **convention**, not a keyword — but always use `self`.

````python
class Dog:
    def __init__(self, name, age):
        self.name = name
        self.age  = age

    def bark(self):
        print(f"{self.name} says: Woof!")

    def info(self):
        print(f"{self.name} is {self.age} year(s) old.")

dog1 = Dog("Rex", 3)
dog2 = Dog("Bella", 5)

dog1.bark()   # Rex says: Woof!
dog2.bark()   # Bella says: Woof!
dog1.info()
````

````python
# Proof that dog1.bark() and Dog.bark(dog1) are identical
dog1.bark()          # normal call
Dog.bark(dog1)       # explicit call — same result
````

## 5. 📦 Instance Attributes vs Class Attributes

| Type | Defined Where | Shared? | Access |
|---|---|---|---|
| **Instance Attribute** | Inside `__init__` via `self.x` | No — each object has its own copy | `obj.x` |
| **Class Attribute** | Directly in the class body | Yes — shared by ALL instances | `Class.x` or `obj.x` |

````python
class Dog:
    # ── Class Attribute ─────────────────────────────
    species = "Canis lupus familiaris"   # shared by all Dog objects
    count   = 0                          # tracks how many dogs were created

    def __init__(self, name, age):
        # ── Instance Attributes ──────────────────────
        self.name = name     # unique to each object
        self.age  = age
        Dog.count += 1       # update class attribute on every creation

dog1 = Dog("Rex", 3)
dog2 = Dog("Bella", 5)
dog3 = Dog("Max", 1)

# Class attributes — same for all
print(Dog.species)        # access via class
print(dog1.species)       # also accessible via instance
print(f"Total dogs created: {Dog.count}")

# Instance attributes — unique per object
print(dog1.name, dog2.name, dog3.name)
````

````python
# ⚠️ Be careful: assigning to an instance shadows the class attribute
dog1.species = "Unknown"     # creates an instance attribute on dog1 only

print(dog1.species)           # Unknown  ← dog1's own copy
print(dog2.species)           # Canis lupus familiaris ← class attribute unchanged
print(Dog.species)            # Canis lupus familiaris ← class attribute unchanged
````

## 6. 🛠️ Instance Methods, Class Methods & Static Methods

Python gives you three kinds of methods inside a class:

```
┌────────────────────┬──────────────────┬──────────────────────────────────┐
│ Method Type        │ First Parameter  │ When to Use                      │
├────────────────────┼──────────────────┼──────────────────────────────────┤
│ Instance Method    │ self             │ Works with instance data         │
│ Class Method       │ cls              │ Works with class-level data      │
│ Static Method      │ (none)           │ Utility — no access to obj/class │
└────────────────────┴──────────────────┴──────────────────────────────────┘
```

````python
class BankAccount:
    bank_name = "PyBank"        # class attribute
    interest_rate = 0.05

    def __init__(self, owner, balance=0):
        self.owner   = owner
        self.balance = balance

    # ── 1. Instance Method ──────────────────────────────────
    def deposit(self, amount):
        """Works with THIS object's balance."""
        if amount > 0:
            self.balance += amount
            print(f"{self.owner}: Deposited {amount}. Balance = {self.balance}")
        else:
            print("Deposit amount must be positive.")

    def withdraw(self, amount):
        if 0 < amount <= self.balance:
            self.balance -= amount
            print(f"{self.owner}: Withdrew {amount}. Balance = {self.balance}")
        else:
            print("Invalid withdrawal amount.")

    # ── 2. Class Method ─────────────────────────────────────
    @classmethod
    def change_interest_rate(cls, new_rate):
        """Changes the rate for ALL accounts."""
        cls.interest_rate = new_rate
        print(f"New interest rate: {cls.interest_rate}")

    @classmethod
    def from_dict(cls, data):
        """Alternative constructor — creates account from a dictionary."""
        return cls(data["owner"], data["balance"])

    # ── 3. Static Method ────────────────────────────────────
    @staticmethod
    def validate_amount(amount):
        """Pure utility — no need for self or cls."""
        return isinstance(amount, (int, float)) and amount > 0


# ----- Usage -----
acc1 = BankAccount("Alice", 1000)
acc2 = BankAccount("Bob",   500)

acc1.deposit(200)
acc1.withdraw(100)

# Class method — call on the class itself
BankAccount.change_interest_rate(0.07)
print(acc1.interest_rate, acc2.interest_rate)  # both updated

# Alternative constructor
data = {"owner": "Charlie", "balance": 750}
acc3 = BankAccount.from_dict(data)
print(acc3.owner, acc3.balance)

# Static method
print(BankAccount.validate_amount(200))    # True
print(BankAccount.validate_amount(-50))    # False
````

## 7. 🖨️ `__str__` and `__repr__`

By default, printing an object shows something like `<__main__.Dog object at 0x...>`. That's not helpful!

| Method | Purpose | Called By |
|---|---|---|
| `__str__` | Human-readable string | `print(obj)`, `str(obj)` |
| `__repr__` | Unambiguous developer string | `repr(obj)`, in interactive shell |

````python
class Book:
    def __init__(self, title, author, pages):
        self.title  = title
        self.author = author
        self.pages  = pages

    def __str__(self):
        """For end users — readable."""
        return f'"{self.title}" by {self.author} ({self.pages} pages)'

    def __repr__(self):
        """For developers — unambiguous, shows how to recreate the object."""
        return f"Book(title={self.title!r}, author={self.author!r}, pages={self.pages})"


b = Book("Clean Code", "Robert C. Martin", 464)

print(b)          # calls __str__
print(str(b))     # calls __str__
print(repr(b))    # calls __repr__
b                 # in a notebook/REPL, calls __repr__
````

## 8. 🔒 Encapsulation — Public, Protected & Private

Encapsulation means **hiding internal details** of an object and only exposing what is necessary.

Python uses **naming conventions** (not strict access modifiers like Java/C++):

| Convention | Syntax | Meaning |
|---|---|---|
| Public | `self.name` | Accessible from anywhere |
| Protected | `self._name` | Convention: "internal use, handle with care" |
| Private | `self.__name` | Name-mangled — harder (but not impossible) to access outside |

````python
class Employee:
    def __init__(self, name, department, salary):
        self.name         = name           # public
        self._department  = department     # protected (by convention)
        self.__salary     = salary         # private (name-mangled)

    def get_info(self):
        # Inside the class, private attributes are accessible normally
        return f"{self.name} | {self._department} | Salary: {self.__salary}"


emp = Employee("Sara", "Engineering", 90000)

print(emp.name)           # ✅ public — accessible
print(emp._department)    # ⚠️  works, but convention says "don't do this outside"
print(emp.get_info())     # ✅ access private data through a method
````

````python
# Trying to access private attribute directly
try:
    print(emp.__salary)
except AttributeError as e:
    print(f"AttributeError: {e}")

# Python name-mangles __salary to _Employee__salary
print(emp._Employee__salary)   # This still works — Python doesn't enforce true privacy
````

## 9. 🏷️ Properties — Getters and Setters with `@property`

Instead of directly exposing private attributes, we use **properties** to control how they are read and written. This is Pythonic encapsulation.

```
  @property          → getter  (called when you READ  obj.attr)
  @attr.setter       → setter  (called when you WRITE obj.attr = value)
  @attr.deleter      → deleter (called when you DELETE del obj.attr)
```

````python
class Circle:
    def __init__(self, radius):
        self.radius = radius    # This will call the setter below!

    @property
    def radius(self):
        """Getter — called when you read circle.radius"""
        return self.__radius

    @radius.setter
    def radius(self, value):
        """Setter — called when you write circle.radius = x"""
        if value < 0:
            raise ValueError("Radius cannot be negative!")
        self.__radius = value

    @property
    def area(self):
        """Read-only computed property — no setter defined."""
        import math
        return round(math.pi * self.__radius ** 2, 4)

    @property
    def circumference(self):
        import math
        return round(2 * math.pi * self.__radius, 4)

    def __str__(self):
        return f"Circle(radius={self.radius}, area={self.area})"


c = Circle(5)
print(c)                  # uses __str__
print(c.radius)           # calls getter
print(c.area)             # computed property
print(c.circumference)

c.radius = 10             # calls setter
print(c.area)             # recalculated automatically
````

````python
# ❌ Validation kicks in via the setter
try:
    c.radius = -3
except ValueError as e:
    print(f"ValueError: {e}")

# ❌ area is read-only — no setter defined
try:
    c.area = 100
except AttributeError as e:
    print(f"AttributeError: {e}")
````

## 10. 🎯 Putting It All Together — Full Example

Let's build a realistic `Student` class that uses everything learned so far.

````python
class Student:
    # Class attribute
    institution = "Python University"
    _student_count = 0

    def __init__(self, name, student_id, age):
        self.name       = name
        self.student_id = student_id
        self.__age      = age          # private
        self.__grades   = {}           # private: {subject: grade}
        Student._student_count += 1

    # ── Property for age ──────────────────────────────────
    @property
    def age(self):
        return self.__age

    @age.setter
    def age(self, value):
        if not (15 <= value <= 100):
            raise ValueError("Age must be between 15 and 100.")
        self.__age = value

    # ── Instance methods ──────────────────────────────────
    def add_grade(self, subject, grade):
        if 0 <= grade <= 100:
            self.__grades[subject] = grade
            print(f"Grade {grade} added for {subject}.")
        else:
            print("Grade must be between 0 and 100.")

    def get_average(self):
        if not self.__grades:
            return 0
        return round(sum(self.__grades.values()) / len(self.__grades), 2)

    def get_report(self):
        print(f"\n{'='*40}")
        print(f"  Student Report")
        print(f"{'='*40}")
        print(f"  Name       : {self.name}")
        print(f"  ID         : {self.student_id}")
        print(f"  Age        : {self.age}")
        print(f"  Institution: {self.institution}")
        print(f"{'─'*40}")
        if self.__grades:
            for subject, grade in self.__grades.items():
                print(f"  {subject:<20} {grade}")
            print(f"{'─'*40}")
            print(f"  Average Grade : {self.get_average()}")
        else:
            print("  No grades recorded.")
        print(f"{'='*40}\n")

    # ── Class method ──────────────────────────────────────
    @classmethod
    def total_students(cls):
        print(f"Total students enrolled: {cls._student_count}")

    # ── Static method ─────────────────────────────────────
    @staticmethod
    def letter_grade(score):
        if score >= 90: return "A"
        elif score >= 80: return "B"
        elif score >= 70: return "C"
        elif score >= 60: return "D"
        else: return "F"

    # ── Dunder methods ────────────────────────────────────
    def __str__(self):
        return f"Student({self.name}, ID={self.student_id})"

    def __repr__(self):
        return f"Student(name={self.name!r}, student_id={self.student_id!r}, age={self.age})"


# ─── Demo ─────────────────────────────────────────────────
s1 = Student("Ahmed Khan", "PU-001", 20)
s2 = Student("Sara Ali",   "PU-002", 22)

s1.add_grade("Mathematics", 88)
s1.add_grade("Physics", 75)
s1.add_grade("Programming", 95)

s1.get_report()

print(f"Average: {s1.get_average()}")
print(f"Letter Grade: {Student.letter_grade(s1.get_average())}")

Student.total_students()

print(str(s1))
print(repr(s1))
````

## 📋 Summary

| Concept | Key Points |
|---|---|
| **Class** | Blueprint defined with `class` keyword |
| **Object / Instance** | Created by calling `ClassName(...)` |
| **`__init__`** | Constructor — sets up instance attributes |
| **`self`** | Reference to the current instance |
| **Instance Attribute** | `self.x` — unique to each object |
| **Class Attribute** | Defined in class body — shared by all instances |
| **Instance Method** | `def method(self)` — works with instance data |
| **Class Method** | `@classmethod` + `cls` — works with class data |
| **Static Method** | `@staticmethod` — utility, no `self`/`cls` |
| **`__str__` / `__repr__`** | Control how objects are printed |
| **Encapsulation** | `_protected`, `__private` — hide internals |
| **`@property`** | Controlled getter/setter without breaking interface |
