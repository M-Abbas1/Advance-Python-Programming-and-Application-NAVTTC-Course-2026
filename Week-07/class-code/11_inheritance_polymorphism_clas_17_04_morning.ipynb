```python
class Shape:
    def area(self):
        return 0
    
    def perimeter(self):
        return 0
    

class Circle(Shape):
    def __init__(self, r):
        self.radius = r

    def area(self):
        a = 3.14 * self.radius ** 2
        return a
    
    def perimeter(self):
        p = 2 * 3.14 * self.radius
        return p
    

class Rectangle(Shape):
    def __init__(self, l, w):
        self.length = l
        self.width = w

    def area(self):
        a = self.length * self.width
        return a
    
    def perimeter(self):
        p = 2 * (self.width + self.length)
        return p
```

```python
c1 = Circle(10)
```

```python
c1.perimeter()
```

```python
r1 = Rectangle(10, 5)
```

```python
r1.perimeter()
```
