```python
class Employee:

    count = 0
    company = "Apple"

    def __init__(self):   
        self.id = None        ## --> instance variables
        self.name = None
        self.salary = 0.0
        self.employement_type = None

    def register_employee(self, id, name, salary, emp_type):
        self.id = id
        self.name = name
        self.salary = salary
        self.employement_type = emp_type
        print("Employee Registered Successfully!")
    
    def show_details(self):
        print("-----Employee Details-----")
        print(f"ID : {self.id}")
        print(f"Name : {self.name}")
        print(f"Salary : {self.salary}")
        print(f"Employement Type : {self.employement_type}")
```

```python
class BankAccount:

    bank_name = "PYTHON BANK LTM"
    branch = "Corvit Branch"

    
    def __init__(self, name, balance):
        self.owner = name
        self.balance = balance

    ## instance method
    def deposit(self, amount):
        if amount > 0:
            self.balance += amount
        else:
            print("Account Balance Must BE Positive")

    ## instance
    def withdraw(self, amount):
        if 0 < amount < self.balance:
            self.balance -= amount
        else:
            print("Something Wrong Happened!")

    ## Class Method
    @classmethod
    def change_branch_name(cls):
        new_branch_name = input("Enter New Branch Name : ")
        cls.branch = new_branch_name
        print("Name changed successfully!")

    @staticmethod
    def validate_amount(amount):
        if isinstance(amount, (int, float)) and amount > 0:
            return True
        else:
            return False



    

        
```
