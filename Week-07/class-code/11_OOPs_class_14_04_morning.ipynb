```python
class ShoppingCart:

    def __init__(self, cart_id, customer_id):
        self.cart_id = cart_id
        self.customer_id = customer_id
        self.items_price = []
        self.items = []
        self.total_price = 0.0

    
    def add_item(self):
        item = input("Select an Item : ")
        price = float(input("Enter item price : "))
        self.items.append(item)
        self.items_price.append(price)
        print(f"Item added, total items : {len(self.items)}")

    def show_items(self):
        print("----- Cart Information -----")
        print(f"Cart ID : {self.cart_id}")
        print(f"Customer ID : {self.customer_id}")
        print(f"Total Items : {len(self.items)}")

        for ind in range(len(self.items)):
            print(self.items[ind], " : ", self.items_price[ind])

        print("-"*30)
        self.calculate_total_price()
        

    def calculate_total_price(self):
        self.total_price = sum(self.items_price)
        print(f"Total Price : {self.total_price}")

```

```python
class ShoppingCart:

    def __init__(self, cart_id, customer_id):
        self.cart_id = cart_id
        self.customer_id = customer_id
        self.items_price = []
        self.items = []
        self.total_price = 0.0

    
    def add_item(self):
        want_to_add = True
        while want_to_add:
            item = input("Select an Item : ")
            price = float(input("Enter item price : "))
            self.items.append(item)
            self.items_price.append(price)
            print(f"Item added, total items : {len(self.items)}")

            choice = input("Do you want to add another item : y/n ")

            if choice=='N' or choice == 'n':
                want_to_add = False
            

    def show_items(self):
        print("----- Cart Information -----")
        print(f"Cart ID : {self.cart_id}")
        print(f"Customer ID : {self.customer_id}")
        print(f"Total Items : {len(self.items)}")

        for ind in range(len(self.items)):
            print(self.items[ind], " : ", self.items_price[ind])

        print("-"*30)
        self.calculate_total_price()
        

    def calculate_total_price(self):
        self.total_price = sum(self.items_price)
        print(f"Total Price : {self.total_price}")

```

```python
shop_cart3 = ShoppingCart(4, 5)
```

```python
shop_cart3.add_item()
```

```python
shop_cart3.show_items()
```

```python
class INT:
    def __init__(self, n):
        self.num = n

    def __str__(self):
        return f"{self.num}"    
    
    def __add__(self, other):
        return self.num + other.num
        
    def __sub__(self, other):
        return self.num - other.num
    
    def __mul__(self, other):
        return self.num * other.num
    
    def __truediv__(self, other):
        if other.num == 0:
            print("Can't divide by Zero")
            return
        return self.num / other.num
```

```python
n1, n2 = INT(40), INT(35)
```
