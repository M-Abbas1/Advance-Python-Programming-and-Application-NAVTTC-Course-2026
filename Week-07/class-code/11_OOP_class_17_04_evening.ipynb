```python
class Car:
    def __init__(self, color, model, brand):
        self.color = color
        self.model = model
        self.brand = brand
        print(id(self))

    def start(self):
        print("Mai chal Pari!  vroom!")

    def brake(self):
        print("back to home, eeessst!")

    def show_details(self):
        print("-----Car Details-----")
        print(f"Brand : {self.brand}")
        print(f"Color : {self.color}")
        print(f"Model : {self.model}")
```
