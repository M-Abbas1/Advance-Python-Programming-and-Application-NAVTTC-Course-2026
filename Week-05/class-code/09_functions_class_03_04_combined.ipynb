```python
def word_analyser(text, key=None):

    words_list = text.split()

    words_info = {}

    for word in words_list:   
        nv = 0
        for char in word:                 
            if char in 'aeiouAEIOU':
                nv += 1
        

        wl = key(word)
        nc = wl - nv

        words_info[word] = [wl, nv, nc]
        
    return words_info
        
```

```python
words_list = ['clear', 'backward', 'something', 'anonymous']
```

```python
words_info = word_analyser(words_list, lambda word: len(word))
```
