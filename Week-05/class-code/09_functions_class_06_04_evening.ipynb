```python
def power(a, b):
    return a**b
```

```python
power(20, 2)
```

```python
# lambda function
```

```python
power = lambda a,b: a**b
```

```python
power(2, 3)
```

```python
power(4, 4)
```

```python
l = ['apples', "mango", "watermelon", "peach"]
```

```python
sorted(l)
```

```python
sorted(l, key=len)
```

```python
ulen = lambda text: len(text)
```

```python
sorted(l, key=ulen)
```

```python
sorted(l, key=lambda text: len(text))
```

```python
students = [
    ('ali', 88),
    ('ahmad', 98),
    ('junaid', 59),
    ('qasim', 79)
]
```

```python
len(students)
```

```python
sorted(students)
```

```python
sorted(students, key=lambda t: t[1])
```
