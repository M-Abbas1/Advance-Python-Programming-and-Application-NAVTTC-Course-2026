```python
def in_stock(product):
    return product['in_stock']
```

```python
list(filter(in_stock, products))
```

```python
available_product  = list(filter(in_stock, products))
```

```python
for product in available_product:
    print(product['name'])
```

```python
for product in list(filter(in_stock, products)):
    print(product['name'])
```

```python
for product in products:
    if product['in_stock'] and product['category'] == 'Electronics':
        print(product['price'])
```

```python
def check(product):
    if product['in_stock'] and product['category'] == "Electronics":
        return True
    else:
        return False
```

```python
list(filter(check, products))
```

```python
electronics_in_stock_items = list(filter(check, products))
```

```python
electronics_in_stock_items
```

```python
for p in electronics_in_stock_items:
    print(p['price'])
```

```python
for p in list(filter(check, products)):
    print(p['price'])
```
