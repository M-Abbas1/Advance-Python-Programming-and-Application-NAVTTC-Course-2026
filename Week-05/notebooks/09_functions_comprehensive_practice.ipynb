# Functions in Python

## Introduction

Functions are one of the most important concepts in programming. Think of a function as a reusable block of code that performs a specific task - like a recipe you can use over and over again. Instead of writing the same code multiple times, you write it once in a function and call it whenever needed.

Imagine you've been writing code to calculate the average of numbers multiple times in your programs. Instead of copying those 3-4 lines of code everywhere, you could create a `calculate_average()` function and just call it with one line!

**Why are functions essential?**
- **Code Reusability** - Write once, use many times
- **Organization** - Break large programs into smaller, manageable pieces
- **Readability** - Well-named functions make code self-documenting
- **Maintainability** - Fix bugs in one place instead of many
- **Testing** - Easy to test individual functions
- **Collaboration** - Team members can work on different functions

## 1. Defining and Calling Functions

The basics of creating and using functions.

### 1.1 Basic Function Definition

```python
# Function syntax:
# def function_name():
#     code block

# Simple function with no parameters
def greet():
    print("Hello, World!")
    print("Welcome to Python Functions!")

# Calling the function
greet()
print()

# Call it multiple times
greet()
greet()
```

### 1.2 Functions with Parameters

```python
# Function with one parameter
def greet_person(name):
    print(f"Hello, {name}!")
    print(f"Nice to meet you, {name}!")

# Calling with different arguments
greet_person("Alice")
print()
greet_person("Bob")

print()
print("=" * 40)

# Function with multiple parameters
def greet_full(first_name, last_name):
    print(f"Hello, {first_name} {last_name}!")

greet_full("Alice", "Smith")
greet_full("Bob", "Johnson")
```

## 2. Return Values

Functions can send back results using return.

### 2.1 Basic Return Statement

```python
# Function that returns a value
def add_numbers(a, b):
    result = a + b
    return result

# Capture the returned value
sum1 = add_numbers(5, 3)
print(f"5 + 3 = {sum1}")

sum2 = add_numbers(10, 20)
print(f"10 + 20 = {sum2}")

# Use directly in expression
total = add_numbers(7, 8) + add_numbers(2, 3)
print(f"Total: {total}")
```

### 2.2 Returning Multiple Values

```python
# Return multiple values (as tuple)
def calculate(a, b):
    addition = a + b
    subtraction = a - b
    multiplication = a * b
    division = a / b if b != 0 else None
    return addition, subtraction, multiplication, division

# Unpack returned values
add, sub, mul, div = calculate(10, 5)
print(f"Addition: {add}")
print(f"Subtraction: {sub}")
print(f"Multiplication: {mul}")
print(f"Division: {div}")

print()

# Or get as tuple
results = calculate(20, 4)
print("All results:", results)
```

### 2.3 Return vs Print

```python
# Function that prints (doesn't return)
def print_sum(a, b):
    print(a + b)

# Function that returns
def return_sum(a, b):
    return a + b

# Using print function
print("Print function:")
result1 = print_sum(5, 3)  # Prints 8
print("Returned value:", result1)  # None

print()

# Using return function
print("Return function:")
result2 = return_sum(5, 3)  # Returns 8, doesn't print
print("Returned value:", result2)  # 8

# Can use in calculations
total = return_sum(5, 3) * 2
print("Total:", total)
```

## 3. Default Parameters

Parameters with default values.

```python
# Function with default parameter
def greet(name, greeting="Hello"):
    print(f"{greeting}, {name}!")

# Call with both arguments
greet("Alice", "Hi")

# Call with only required argument (uses default)
greet("Bob")

# Different greetings
greet("Charlie", "Good morning")
greet("Diana")

print()

# Multiple default parameters
def create_profile(name, age=18, city="Unknown"):
    print(f"Name: {name}")
    print(f"Age: {age}")
    print(f"City: {city}")
    print()

create_profile("Alice", 25, "Boston")
create_profile("Bob", 30)
create_profile("Charlie")
```

## 4. Keyword Arguments

Calling functions with named arguments.

```python
def create_user(username, email, age, country):
    print(f"Username: {username}")
    print(f"Email: {email}")
    print(f"Age: {age}")
    print(f"Country: {country}")
    print()

# Positional arguments (order matters)
create_user("alice123", "alice@email.com", 25, "USA")

# Keyword arguments (order doesn't matter)
create_user(email="bob@email.com", username="bob456", country="UK", age=30)

# Mix positional and keyword (positional must come first)
create_user("charlie", "charlie@email.com", country="Canada", age=28)
```

## 5. Variable-Length Arguments

*args and **kwargs for flexible functions.

### 5.1 *args - Variable Positional Arguments

```python
# *args allows any number of positional arguments
def sum_all(*numbers):
    print(f"Received {len(numbers)} numbers")
    total = sum(numbers)
    return total

# Call with different number of arguments
print("Sum of 1, 2, 3:", sum_all(1, 2, 3))
print("Sum of 10, 20:", sum_all(10, 20))
print("Sum of 5, 10, 15, 20, 25:", sum_all(5, 10, 15, 20, 25))

print()

# Accessing individual arguments
def print_all(*items):
    print("Items received:")
    for i, item in enumerate(items, 1):
        print(f"  {i}. {item}")

print_all("apple", "banana", "cherry")
print()
print_all(1, 2, 3, 4, 5)
```

### 5.2 **kwargs - Variable Keyword Arguments

```python
# **kwargs allows any number of keyword arguments
def print_info(**info):
    print("Information received:")
    for key, value in info.items():
        print(f"  {key}: {value}")
    print()

# Call with different keyword arguments
print_info(name="Alice", age=25, city="Boston")
print_info(product="Laptop", price=999.99, brand="TechCo")

# Combining regular parameters with **kwargs
def create_student(name, **details):
    print(f"Student: {name}")
    for key, value in details.items():
        print(f"  {key}: {value}")
    print()

create_student("Bob", age=20, major="CS", gpa=3.8)
create_student("Charlie", age=22, major="Math")
```

### 5.3 Combining *args and **kwargs

```python
# Function accepting both *args and **kwargs
def flexible_function(*args, **kwargs):
    print("Positional arguments:")
    for arg in args:
        print(f"  - {arg}")
    
    print("\nKeyword arguments:")
    for key, value in kwargs.items():
        print(f"  {key} = {value}")
    print()

flexible_function(1, 2, 3, name="Alice", age=25)
flexible_function("apple", "banana", color="yellow", count=5)
```

## 6. Scope - Local vs Global Variables

### 6.1 Local Scope

```python
# Variables inside function are local
def my_function():
    local_var = 10  # Local variable
    print("Inside function:", local_var)

my_function()

# Cannot access local variable outside function
# print(local_var)  # NameError!
```

### 6.2 Global Scope

```python
# Global variable
global_var = 100

def access_global():
    # Can read global variable
    print("Inside function:", global_var)

access_global()
print("Outside function:", global_var)
```

### 6.3 Modifying Global Variables

```python
counter = 0

def increment():
    global counter  # Declare as global to modify
    counter += 1
    print(f"Counter: {counter}")

print("Initial:", counter)
increment()
increment()
increment()
print("Final:", counter)
```

### 6.4 Local Variable Shadowing

```python
x = 10  # Global

def test():
    x = 20  # Local (shadows global)
    print("Inside function:", x)

print("Before function:", x)
test()
print("After function:", x)  # Global unchanged
```

## 7. Lambda Functions

Anonymous functions for simple operations.

### 7.1 Basic Lambda Functions

```python
# Regular function
def square(x):
    return x ** 2

# Lambda equivalent
square_lambda = lambda x: x ** 2

print("Regular function:", square(5))
print("Lambda function:", square_lambda(5))

print()

# Lambda with multiple parameters
add = lambda a, b: a + b
print("Add:", add(3, 7))

multiply = lambda a, b: a * b
print("Multiply:", multiply(4, 5))
```

### 7.2 Lambda with Built-in Functions

```python
# Using lambda with sorted()
students = [
    ("Alice", 85),
    ("Bob", 92),
    ("Charlie", 78),
    ("Diana", 95)
]

# Sort by grade (second element)
sorted_students = sorted(students, key=lambda x: x[1], reverse=True)
print("Sorted by grade:")
for name, grade in sorted_students:
    print(f"  {name}: {grade}")

print()

# Using lambda with map()
numbers = [1, 2, 3, 4, 5]
squared = list(map(lambda x: x ** 2, numbers))
print("Squared:", squared)

# Using lambda with filter()
even_numbers = list(filter(lambda x: x % 2 == 0, numbers))
print("Even numbers:", even_numbers)
```

## 8. Docstrings and Documentation

```python
def calculate_area(length, width):
    """
    Calculate the area of a rectangle.
    
    Parameters:
        length (float): The length of the rectangle
        width (float): The width of the rectangle
    
    Returns:
        float: The area of the rectangle
    
    Example:
        >>> calculate_area(5, 3)
        15
    """
    return length * width

# Access docstring
print(calculate_area.__doc__)

# Use the function
area = calculate_area(10, 5)
print(f"\nArea: {area}")
```

## 9. Practical Examples

### 9.1 Input Validation Function

```python
def get_positive_number(prompt):
    """
    Get a positive number from user with validation.
    """
    while True:
        try:
            value = float(input(prompt))
            if value > 0:
                return value
            else:
                print("Please enter a positive number.")
        except ValueError:
            print("Invalid input. Please enter a number.")

# Example usage (commented out for notebook)
# num = get_positive_number("Enter a positive number: ")
# print(f"You entered: {num}")
```

### 9.2 Grade Calculator

```python
def calculate_grade(score):
    """
    Convert numeric score to letter grade.
    """
    if score >= 90:
        return 'A'
    elif score >= 80:
        return 'B'
    elif score >= 70:
        return 'C'
    elif score >= 60:
        return 'D'
    else:
        return 'F'

def calculate_gpa(grade):
    """
    Convert letter grade to GPA.
    """
    grade_points = {'A': 4.0, 'B': 3.0, 'C': 2.0, 'D': 1.0, 'F': 0.0}
    return grade_points.get(grade, 0.0)

# Test
scores = [95, 87, 72, 65, 58]
print("Score\tGrade\tGPA")
print("-" * 25)
for score in scores:
    grade = calculate_grade(score)
    gpa = calculate_gpa(grade)
    print(f"{score}\t{grade}\t{gpa}")
```

### 9.3 Statistics Functions

```python
def calculate_statistics(numbers):
    """
    Calculate mean, median, and mode of a list.
    """
    # Mean
    mean = sum(numbers) / len(numbers)
    
    # Median
    sorted_nums = sorted(numbers)
    n = len(sorted_nums)
    if n % 2 == 0:
        median = (sorted_nums[n//2 - 1] + sorted_nums[n//2]) / 2
    else:
        median = sorted_nums[n//2]
    
    # Range
    range_val = max(numbers) - min(numbers)
    
    return mean, median, range_val

# Test
data = [85, 90, 78, 92, 88, 85, 95]
mean, median, range_val = calculate_statistics(data)

print("Data:", data)
print(f"Mean: {mean:.2f}")
print(f"Median: {median}")
print(f"Range: {range_val}")
```
