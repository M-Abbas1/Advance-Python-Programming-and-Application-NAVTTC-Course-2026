## Key Tips

### Tuples
```python
# Single element needs comma
single = (5,)  # Tuple
not_tuple = (5)  # Integer

# Unpacking
x, y = (10, 20)
first, *rest = (1, 2, 3, 4)

# Swapping
a, b = b, a
```

### Sets
```python
# Empty set
s = set()  # NOT {}

# Remove duplicates
unique = list(set(my_list))

# Operations
union = set1 | set2
intersection = set1 & set2
difference = set1 - set2
sym_diff = set1 ^ set2

# Safe removal
s.discard(item)  # No error if missing
```

### When to Use
- **Tuple:** Coordinates, RGB, fixed records, dict keys
- **Set:** Unique items, membership testing, removing duplicates
- **List:** Ordered, mutable collection
- **Dict:** Key-value mappings
