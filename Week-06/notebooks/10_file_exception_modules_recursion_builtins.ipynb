## 1. File Handling

Python makes it easy to read data from files and write results back to disk.
Understanding file I/O is essential for any real-world programme.

### 1.1 Writing to a File (write & writelines)

````python
# write() – writes a single string; does NOT add a newline automatically
with open("demo.txt", "w") as f:
    f.write("Hello, File!\n")
    f.write("Second line\n")
    f.write("Third line\n")

print("File written successfully.")

# writelines() – writes a list of strings (no automatic newlines either)
lines = ["Apple\n", "Banana\n", "Cherry\n"]
with open("fruits.txt", "w") as f:
    f.writelines(lines)

print("Fruits file written.")
````

### 1.2 Reading Files (read, readline, readlines)

````python
# read() – reads the ENTIRE file as one big string
with open("demo.txt", "r") as f:
    content = f.read()
print("=== read() ===")
print(content)

# readline() – reads ONE line at a time; returns '' at end of file
print("=== readline() ===")
with open("demo.txt", "r") as f:
    line1 = f.readline()
    line2 = f.readline()
print(repr(line1))
print(repr(line2))

# readlines() – reads ALL lines into a list
print("=== readlines() ===")
with open("demo.txt", "r") as f:
    all_lines = f.readlines()
print(all_lines)
````

````python
# Iterating over a file line by line (memory-efficient for large files)
print("=== Iterating line by line ===")
with open("fruits.txt", "r") as f:
    for i, line in enumerate(f, 1):
        print(f"Line {i}: {line.strip()}")
````

### 1.3 File Modes

| Mode | Meaning |
|------|---------|
| `r`  | Read (default). Error if file missing. |
| `w`  | Write. Creates or **overwrites** file. |
| `a`  | Append. Creates if missing; never overwrites. |
| `r+` | Read **and** write. File must exist. |
| `w+` | Write and read. Creates or overwrites. |
| `a+` | Append and read. |
| `rb` / `wb` | Binary read / write. |

````python
# Append mode – adds to the end without destroying existing content
with open("demo.txt", "a") as f:
    f.write("Appended line\n")

with open("demo.txt", "r") as f:
    print("After append:")
    print(f.read())

# r+ mode – read AND write without truncating
with open("demo.txt", "r+") as f:
    original = f.read()
    print(f"Characters in file: {len(original)}")
    f.write("Extra via r+\n")   # appends at current position (end after read)

# Binary mode – useful for images, audio, etc.
data = bytes([72, 101, 108, 108, 111])   # "Hello" in bytes
with open("binary.bin", "wb") as f:
    f.write(data)

with open("binary.bin", "rb") as f:
    raw = f.read()
print(f"Binary data: {raw}")
print(f"Decoded: {raw.decode()}")
````

### 1.4 Context Managers – the `with` Statement

The `with` statement automatically closes the file even if an exception occurs.
Always prefer `with` over manual `f.open()` / `f.close()` pairs.

````python
# WITHOUT with (old style – risky if exception occurs between open and close)
f = open("demo.txt", "r")
content = f.read()
f.close()   # easy to forget!
print("Manual close – first 30 chars:", content[:30])

# WITH context manager (recommended)
with open("demo.txt", "r") as f:
    content = f.read()
# File is automatically closed here, even if an exception was raised
print("Context manager – file closed:", f.closed)

# Nesting context managers (Python 3.1+)
with open("demo.txt", "r") as src, open("demo_copy.txt", "w") as dst:
    for line in src:
        dst.write(line)
print("File copied successfully.")
````

````python
# tell() and seek() – navigating within a file
with open("demo.txt", "r") as f:
    print("Initial position:", f.tell())
    first_line = f.readline()
    print(f"After reading first line, position: {f.tell()}")
    
    # seek(0) rewinds to the beginning
    f.seek(0)
    print(f"After seek(0), position: {f.tell()}")
    print("Re-reading first line:", f.readline().strip())
````

## 2. Exception Handling

Errors are inevitable. Exception handling lets your programme respond to them
gracefully instead of crashing.

### 2.1 Basic try-except Block

````python
# Without exception handling – crashes on bad input
# result = 10 / 0   # ZeroDivisionError

# With exception handling
try:
    result = 10 / 0
except ZeroDivisionError:
    print("Error: cannot divide by zero!")

print("Programme continues normally after the except block.")

# Another example – bad type conversion
try:
    number = int("hello")
except ValueError as e:
    print(f"ValueError caught: {e}")

print("Still running!")
````

### 2.2 Catching Multiple Exceptions

````python
# Separate except clauses for different error types
def safe_divide(a, b):
    try:
        result = a / b
        return result
    except ZeroDivisionError:
        print("  Cannot divide by zero!")
        return None
    except TypeError:
        print("  Both arguments must be numbers!")
        return None

print(safe_divide(10, 2))
print(safe_divide(10, 0))
print(safe_divide(10, "x"))
````

````python
# Catching multiple exceptions in ONE clause
def parse_index(data, index_str):
    try:
        index = int(index_str)
        return data[index]
    except (ValueError, IndexError) as e:
        print(f"Error ({type(e).__name__}): {e}")
        return None

items = ["a", "b", "c"]
print(parse_index(items, "1"))    # fine
print(parse_index(items, "ten")) # ValueError
print(parse_index(items, "99"))  # IndexError
````

### 2.3 `else` and `finally` Clauses

````python
# else   – runs only when NO exception was raised
# finally – ALWAYS runs (cleanup code goes here)

def open_file(filename):
    try:
        f = open(filename, "r")
    except FileNotFoundError:
        print(f"  File '{filename}' not found.")
    else:
        content = f.read()
        f.close()
        print(f"  File read successfully: {len(content)} characters.")
    finally:
        print("  (finally block always executes)")

open_file("demo.txt")       # exists
print()
open_file("missing.txt")    # does not exist
````

````python
# Real-world pattern: resource cleanup with finally
def process_data(filename):
    f = None
    try:
        f = open(filename, "r")
        data = f.read()
        # Simulate processing error
        value = int(data.strip())
        result = 100 / value
        return result
    except FileNotFoundError:
        print("File missing!")
    except ValueError:
        print("File does not contain a number!")
    except ZeroDivisionError:
        print("Number in file is zero – cannot divide!")
    finally:
        if f:
            f.close()
            print("File closed in finally.")

# Create a test file
with open("number.txt", "w") as f:
    f.write("5")

print(process_data("number.txt"))
````

### 2.4 Raising Exceptions

````python
# raise – manually trigger an exception
def set_age(age):
    if not isinstance(age, int):
        raise TypeError(f"Age must be an integer, got {type(age).__name__}")
    if age < 0:
        raise ValueError(f"Age cannot be negative: {age}")
    if age > 150:
        raise ValueError(f"Age {age} is unrealistically large")
    return age

# Valid
print(set_age(25))

# Invalid
for bad in [-5, 200, "old"]:
    try:
        print(set_age(bad))
    except (TypeError, ValueError) as e:
        print(f"  Caught: {e}")
````

````python
# Re-raising exceptions (add context, then propagate)
def load_config(path):
    try:
        with open(path) as f:
            import json
            return json.load(f)
    except FileNotFoundError:
        raise FileNotFoundError(f"Config file not found at: {path}") from None
    except json.JSONDecodeError as e:
        raise ValueError(f"Config file is invalid JSON: {e}") from e

try:
    load_config("nonexistent_config.json")
except FileNotFoundError as e:
    print(f"Caught re-raised error: {e}")
````

### 2.5 Custom Exceptions

````python
# Custom exceptions let you create domain-specific error types

class BankError(Exception):
    """Base exception for all banking errors."""
    pass

class InsufficientFundsError(BankError):
    """Raised when a withdrawal exceeds the available balance."""
    def __init__(self, amount, balance):
        self.amount = amount
        self.balance = balance
        super().__init__(
            f"Cannot withdraw £{amount:.2f} – balance is only £{balance:.2f}"
        )

class NegativeAmountError(BankError):
    """Raised when a negative amount is supplied."""
    pass


class BankAccount:
    def __init__(self, owner, balance=0):
        self.owner = owner
        self.balance = balance

    def deposit(self, amount):
        if amount <= 0:
            raise NegativeAmountError("Deposit amount must be positive")
        self.balance += amount
        print(f"  Deposited £{amount:.2f}. New balance: £{self.balance:.2f}")

    def withdraw(self, amount):
        if amount <= 0:
            raise NegativeAmountError("Withdrawal amount must be positive")
        if amount > self.balance:
            raise InsufficientFundsError(amount, self.balance)
        self.balance -= amount
        print(f"  Withdrew £{amount:.2f}. New balance: £{self.balance:.2f}")


account = BankAccount("Alice", 100)
account.deposit(50)

for attempt in [30, 200, -10]:
    try:
        account.withdraw(attempt)
    except InsufficientFundsError as e:
        print(f"  InsufficientFunds: {e}")
    except NegativeAmountError as e:
        print(f"  NegativeAmount: {e}")
````

## 4. Recursion
A technique where a function calls itself, directly or indirectly, to solve a problem by breaking it down into smaller, self-similar subproblems.
A function is **recursive** when it calls itself.
Every recursive solution needs:
1. A **base case** – the condition that stops the recursion.
2. A **recursive case** – the step that moves toward the base case.

### 4.1 Your First Recursive Function

````python
def countdown(n):
    """Count down from n to 0."""
    if n <= 0:          # base case
        print("Go!")
        return
    print(n)
    countdown(n - 1)   # recursive case – n decreases each time

countdown(5)
````

````python
# Visualise the call stack
def factorial_verbose(n, depth=0):
    indent = "  " * depth
    print(f"{indent}factorial({n}) called")
    
    if n == 0:          # base case
        print(f"{indent}returning 1 (base case)")
        return 1
    
    result = n * factorial_verbose(n - 1, depth + 1)
    print(f"{indent}returning {n} * ... = {result}")
    return result

print(factorial_verbose(4))
````

### 4.2 Classic Recursive Problems

````python
# ── Factorial ──────────────────────────────────────────────────
def factorial(n):
    if n == 0:
        return 1
    return n * factorial(n - 1)

for i in range(8):
    print(f"{i}! = {factorial(i)}")
````

````python
# ── Fibonacci ──────────────────────────────────────────────────
def fibonacci(n):
    """Return the nth Fibonacci number (0-indexed)."""
    if n <= 1:          # base cases
        return n
    return fibonacci(n - 1) + fibonacci(n - 2)

sequence = [fibonacci(i) for i in range(12)]
print("Fibonacci sequence:", sequence)
````

````python
# ── Binary Search (recursive) ──────────────────────────────────
def binary_search(arr, target, low=0, high=None):
    if high is None:
        high = len(arr) - 1

    if low > high:          # base case – not found
        return -1

    mid = (low + high) // 2

    if arr[mid] == target:  # base case – found
        return mid
    elif arr[mid] < target:
        return binary_search(arr, target, mid + 1, high)
    else:
        return binary_search(arr, target, low, mid - 1)

sorted_list = [2, 5, 8, 12, 16, 23, 38, 45, 56, 72, 91]
print("List:", sorted_list)

for val in [23, 1, 91, 50]:
    idx = binary_search(sorted_list, val)
    if idx != -1:
        print(f"  {val} found at index {idx}")
    else:
        print(f"  {val} not found")
````

````python
# ── Tower of Hanoi ─────────────────────────────────────────────
def hanoi(n, source="A", target="C", auxiliary="B"):
    """Print moves to solve Tower of Hanoi with n discs."""
    if n == 1:
        print(f"  Move disc 1 from {source} → {target}")
        return
    hanoi(n - 1, source, auxiliary, target)
    print(f"  Move disc {n} from {source} → {target}")
    hanoi(n - 1, auxiliary, target, source)

print("Tower of Hanoi – 3 discs:")
hanoi(3)
````

### 4.3 Recursion vs Iteration

````python
import time

# ── Sum of list – iterative vs recursive ───────────────────────

def sum_iterative(lst):
    total = 0
    for x in lst:
        total += x
    return total

def sum_recursive(lst):
    if not lst:
        return 0
    return lst[0] + sum_recursive(lst[1:])

data = list(range(1, 101))

t0 = time.perf_counter()
result_iter = sum_iterative(data)
t1 = time.perf_counter()

result_rec = sum_recursive(data)
t2 = time.perf_counter()

print(f"Iterative result: {result_iter}  | time: {(t1-t0)*1e6:.1f} µs")
print(f"Recursive result: {result_rec}  | time: {(t2-t1)*1e6:.1f} µs")
print()
print("Key differences:")
print("  • Recursion: elegant, mirrors mathematical definitions")
print("  • Iteration: faster, no stack-overflow risk, preferred for large inputs")
````

````python
# Memoisation – dramatically speeds up naive Fibonacci
from functools import lru_cache
import time

def fib_slow(n):
    if n <= 1:
        return n
    return fib_slow(n-1) + fib_slow(n-2)

@lru_cache(maxsize=None)
def fib_fast(n):
    if n <= 1:
        return n
    return fib_fast(n-1) + fib_fast(n-2)

n = 35
t0 = time.perf_counter(); slow = fib_slow(n); t1 = time.perf_counter()
t2 = time.perf_counter(); fast = fib_fast(n); t3 = time.perf_counter()

print(f"fib({n}) = {fast}")
print(f"Without memoisation: {(t1-t0)*1000:.1f} ms")
print(f"With @lru_cache:     {(t3-t2)*1000:.4f} ms")
print(f"Speedup: ~{(t1-t0)/(t3-t2+1e-12):.0f}×")
````

## 5. Built-in Functions

Python's built-in functions let you write expressive, concise code
without importing anything.

### 5.1 `map()` – Apply a Function to Every Item

````python
# map(function, iterable) → lazy iterator
numbers = [1, 2, 3, 4, 5]

# Double every number
doubled = list(map(lambda x: x * 2, numbers))
print("Doubled:", doubled)

# Convert strings to ints
str_nums = ["10", "20", "30", "40"]
ints = list(map(int, str_nums))
print("Converted:", ints)

# Apply to multiple iterables
a = [1, 2, 3]
b = [10, 20, 30]
sums = list(map(lambda x, y: x + y, a, b))
print("Element-wise sums:", sums)

# map with a defined function
def celsius_to_fahrenheit(c):
    return round(c * 9/5 + 32, 1)

temps_c = [0, 20, 37, 100]
temps_f = list(map(celsius_to_fahrenheit, temps_c))
print("Celsius:", temps_c)
print("Fahrenheit:", temps_f)
````

### 5.2 `filter()` – Keep Items That Pass a Test

````python
# filter(function, iterable) → lazy iterator with only truthy results

numbers = range(1, 21)

evens     = list(filter(lambda x: x % 2 == 0, numbers))
odds      = list(filter(lambda x: x % 2 != 0, numbers))
over_ten  = list(filter(lambda x: x > 10, numbers))

print("Evens:", evens)
print("Odds:", odds)
print("Over 10:", over_ten)

# Filter strings
words = ["apple", "banana", "cherry", "avocado", "blueberry", "apricot"]
a_words = list(filter(lambda w: w.startswith("a"), words))
long_words = list(filter(lambda w: len(w) > 6, words))
print("\nStarting with 'a':", a_words)
print("Longer than 6 chars:", long_words)

# Combining map and filter
numbers = range(1, 11)
result = list(map(lambda x: x**2, filter(lambda x: x % 2 == 0, numbers)))
print("\nSquares of even numbers 1-10:", result)
````

### 5.3 `reduce()` – Fold a Sequence to a Single Value

````python
from functools import reduce

numbers = [1, 2, 3, 4, 5]

# Sum (same as sum())
total = reduce(lambda acc, x: acc + x, numbers)
print("Sum:", total)

# Product
product = reduce(lambda acc, x: acc * x, numbers)
print("Product:", product)

# Maximum (same as max())
maximum = reduce(lambda a, b: a if a > b else b, numbers)
print("Max:", maximum)

# Flatten a list of lists
nested = [[1, 2], [3, 4], [5, 6]]
flat = reduce(lambda acc, lst: acc + lst, nested)
print("Flattened:", flat)

# Build a string
words = ["Python", "is", "fun"]
sentence = reduce(lambda s, w: s + " " + w, words)
print("Sentence:", sentence)
````

### 5.4 `zip()` and `enumerate()`

````python
# zip() – pair up iterables element by element
names  = ["Alice", "Bob", "Carol"]
scores = [85, 92, 78]
grades = ["B", "A", "C"]

# Zip two lists
paired = list(zip(names, scores))
print("Zipped:", paired)

# Zip three lists
tripled = list(zip(names, scores, grades))
for name, score, grade in tripled:
    print(f"  {name}: {score} ({grade})")

print()

# Unzip with *
zipped = [(1, "a"), (2, "b"), (3, "c")]
nums, letters = zip(*zipped)
print("Unzipped nums:", nums)
print("Unzipped letters:", letters)

# zip stops at the shortest iterable
short = [1, 2]
long  = [10, 20, 30, 40]
print("zip short+long:", list(zip(short, long)))
````

````python
# enumerate() – add an index counter to any iterable
fruits = ["apple", "banana", "cherry", "date"]

# Default start=0
for i, fruit in enumerate(fruits):
    print(f"  {i}: {fruit}")

print()

# Start from 1
for i, fruit in enumerate(fruits, start=1):
    print(f"  {i}. {fruit}")

print()

# Practical: find index of items matching a condition
numbers = [15, 3, 42, 8, 23, 16, 4]
large = [(i, n) for i, n in enumerate(numbers) if n > 10]
print("Indices of numbers > 10:", large)
````

### 5.5 `any()` and `all()`

````python
# any(iterable) – True if AT LEAST ONE element is truthy
# all(iterable) – True if ALL elements are truthy

scores = [88, 92, 45, 76, 61]

print("=== Scores:", scores, "===")
print("Any score > 90?", any(s > 90 for s in scores))
print("All scores > 50?", all(s > 50 for s in scores))
print("All scores > 40?", all(s > 40 for s in scores))
print("Any score < 50?", any(s < 50 for s in scores))

print()

# Checking user permissions
permissions = {"read": True, "write": True, "delete": False}
print("Can do anything?", any(permissions.values()))
print("Can do everything?", all(permissions.values()))
print("Can read and write?", all([permissions["read"], permissions["write"]]))

print()

# Validating a list
def is_valid_email(email):
    return "@" in email and "." in email

emails = ["user@example.com", "bad-email", "another@test.org"]
print("All valid?", all(is_valid_email(e) for e in emails))
print("Any valid?", any(is_valid_email(e) for e in emails))
````

### 5.6 `sorted()` and `reversed()`

````python
# sorted() returns a NEW sorted list; original unchanged
# list.sort() sorts IN-PLACE and returns None

numbers = [3, 1, 4, 1, 5, 9, 2, 6, 5, 3]
print("Original:", numbers)
print("sorted():", sorted(numbers))
print("sorted(reverse=True):", sorted(numbers, reverse=True))
print("Original unchanged:", numbers)

print()

# Sort with a key function
words = ["banana", "apple", "cherry", "date", "elderberry"]
print("By length:", sorted(words, key=len))
print("Alphabetical:", sorted(words))
print("By last char:", sorted(words, key=lambda w: w[-1]))

print()

# Sort complex objects
students = [
    {"name": "Alice", "gpa": 3.8},
    {"name": "Bob",   "gpa": 3.5},
    {"name": "Carol", "gpa": 3.9},
    {"name": "Dave",  "gpa": 3.5},
]
by_gpa = sorted(students, key=lambda s: s["gpa"], reverse=True)
for s in by_gpa:
    print(f"  {s['name']}: {s['gpa']}")
````

````python
# reversed() – iterate in reverse; works on sequences
numbers = [1, 2, 3, 4, 5]
print("reversed list:", list(reversed(numbers)))

for n in reversed(numbers):
    print(n, end=" ")
print()

# Reverse a string
text = "Hello, Python!"
reversed_text = "".join(reversed(text))
print("Reversed string:", reversed_text)

# Compare with slicing
print("Slice reverse:", text[::-1])

# Reverse range
print("Reversed range:", list(reversed(range(5))))
````

### 5.7 Combining Built-in Functions

````python
# Real-world example – student report card
students = [
    {"name": "Alice",   "scores": [88, 92, 79, 95, 84]},
    {"name": "Bob",     "scores": [72, 68, 74, 70, 76]},
    {"name": "Carol",   "scores": [95, 98, 92, 96, 99]},
    {"name": "Dave",    "scores": [55, 62, 48, 59, 61]},
    {"name": "Eve",     "scores": [83, 87, 91, 79, 88]},
]

# Calculate average for each student using map + sum/len
def average(scores):
    return sum(scores) / len(scores)

with_avg = list(map(lambda s: {**s, "avg": round(average(s["scores"]), 1)}, students))

# Filter only passing students (avg >= 70)
passing = list(filter(lambda s: s["avg"] >= 70, with_avg))

# Sort by average descending
ranked = sorted(passing, key=lambda s: s["avg"], reverse=True)

# Print report using enumerate
print("=== Passing Students (ranked) ===")
for rank, student in enumerate(ranked, 1):
    passed_all = all(sc >= 50 for sc in student["scores"])
    flag = "✓" if passed_all else "⚠ has a low score"
    print(f"  {rank}. {student['name']:<8} avg={student['avg']}  {flag}")

# Check if anyone got a perfect 100
scores_flat = [sc for s in students for sc in s["scores"]]
print(f"\nAny perfect 100? {any(sc == 100 for sc in scores_flat)}")
print(f"All scores ≥ 50?  {all(sc >= 50 for sc in scores_flat)}")
````

### Exercise 2 – Exception Handling

Create a function `safe_calculator(a, op, b)` that:
- Supports `+`, `-`, `*`, `/`
- Raises `ValueError` for an unknown operator
- Raises `ZeroDivisionError` (with a descriptive message) for division by zero
- Returns the result for valid inputs
- Test it with at least 5 cases including error cases.

````python
# Your solution here
````

**Solution – Exercise 2**

````python
class UnknownOperatorError(ValueError):
    pass

def safe_calculator(a, op, b):
    if op not in ("+", "-", "*", "/"):
        raise UnknownOperatorError(f"Unknown operator: '{op}'. Use +, -, * or /")
    if op == "/" and b == 0:
        raise ZeroDivisionError(f"Division by zero: {a} / 0 is undefined")
    ops = {"+": a + b, "-": a - b, "*": a * b, "/": a / b}
    return ops[op]

tests = [
    (10, "+", 5),
    (10, "-", 3),
    (4,  "*", 7),
    (9,  "/", 3),
    (5,  "/", 0),
    (8,  "%", 3),
]

for a, op, b in tests:
    try:
        result = safe_calculator(a, op, b)
        print(f"  {a} {op} {b} = {result}")
    except ZeroDivisionError as e:
        print(f"  ZeroDivisionError: {e}")
    except UnknownOperatorError as e:
        print(f"  UnknownOperatorError: {e}")
````

````python
module = '''
def reverse_string(s):
    return s[::-1]

def is_palindrome(s):
    clean = s.lower().replace(" ", "")
    return clean == clean[::-1]

def word_count(text):
    freq = {}
    for word in text.lower().split():
        word = word.strip(".,!?;:")
        freq[word] = freq.get(word, 0) + 1
    return freq

def title_case(text):
    return " ".join(word[0].upper() + word[1:].lower() for word in text.split())
'''

with open("string_utils.py", "w") as f:
    f.write(module)

import string_utils as su

print(su.reverse_string("Hello, World!"))
print(su.is_palindrome("racecar"))
print(su.is_palindrome("A man a plan a canal Panama"))
print(su.is_palindrome("hello"))
print(su.word_count("the cat sat on the mat and the cat sat"))
print(su.title_case("the quick BROWN fox"))
````

### Exercise 4 – Recursion

Implement the following recursive functions:
1. `power(base, exp)` – compute base^exp without using `**`
2. `flatten(lst)` – flatten a deeply nested list
3. `sum_digits(n)` – return the sum of the digits of integer n

````python
# Your solution here
````

**Solution – Exercise 4**

````python
def power(base, exp):
    """Compute base^exp recursively."""
    if exp == 0:
        return 1
    if exp < 0:
        return 1 / power(base, -exp)
    return base * power(base, exp - 1)

def flatten(lst):
    """Recursively flatten a nested list."""
    result = []
    for item in lst:
        if isinstance(item, list):
            result.extend(flatten(item))
        else:
            result.append(item)
    return result

def sum_digits(n):
    """Return sum of digits of n."""
    n = abs(n)
    if n < 10:
        return n
    return (n % 10) + sum_digits(n // 10)

# Tests
print("power(2, 10):", power(2, 10))
print("power(3, 0):", power(3, 0))
print("power(2, -3):", power(2, -3))

nested = [1, [2, [3, [4, 5]], 6], 7, [8, 9]]
print("flatten:", flatten(nested))

for n in [0, 7, 123, 9999, 12345]:
    print(f"sum_digits({n}) = {sum_digits(n)}")
````
