```python
# ── Tower of Hanoi ─────────────────────────────────────────────
def hanoi(n, source="A", target="C", auxiliary="B"):
    """Print moves to solve Tower of Hanoi with n discs."""
    if n == 1:
        print(f"  Move disc 1 from {source} → {target}")
        return
    hanoi(n - 1, source, auxiliary, target)
    print(f"  Move disc {n} from {source} → {target}")
    hanoi(n - 1, auxiliary, target, source)

print("Tower of Hanoi – 3 discs:")
hanoi(5)
```

```python
def cal():

    try:
        a = float(input("Enter number 1 : "))
        b = float(input("Enter number 2 : "))
    except ValueError:
        print("Enter a valid input")
    else:     
        op = input("Enter an operator : ")
        res = 0

        if op == '+':
            res = a + b
        elif op == '-':
            res = a - b
        elif op == '*':
            res = a * b
        elif op == '/':
            try:
                res = a / b
            except ZeroDivisionError:
                print("can't divide by zero")

        return res
    
```

```python
student_list = []

for st_key, st_value in students.items():
    class_no = st_key
    name = st_value['name']
    age = str(st_value['age'])
    dept = st_value['department']
    city = st_value['city']
    course1 = st_value['courses'][0]
    course2 = st_value['courses'][1]
    course3 = st_value['courses'][2]
    course1_grades = str(st_value['grades'][course1])
    course2_grades = str(st_value['grades'][course2])
    course3_grades = str(st_value['grades'][course3])
    current_student = [class_no, name, age, dept, city, course1, course1_grades, course2, course2_grades, course3, course3_grades]

    current_student_str = ",".join(current_student)

    student_list.append(current_student_str)
```

```python
student_list
```

```python
with open('student_data.txt', 'w') as file:
    file.write("Class No,Name,Age,Department,City,Course1, grade1,Cours2, grade2, Course3, grade3")
    for line in student_list:
        file.write(line+"\n")
```
