```python
def countdown(n):

    if n <= 0:
        print("Go!")
        return 
    
    print(n)
    countdown(n-1)


countdown(5)
```

```python
def factorial(n):
    if n == 1:
        return 1
    print(n)

    return n * factorial(n-1)   

factorial(5)
```
