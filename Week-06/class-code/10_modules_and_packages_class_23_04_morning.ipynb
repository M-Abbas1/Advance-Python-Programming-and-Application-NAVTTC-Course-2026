```python
import random
import math
import datetime
import sys
import os

# Get difficulty from command line
if len(sys.argv) > 1:
    difficulty = sys.argv[1]
else:
    difficulty = "easy"

# Difficulty settings
if difficulty == "easy":
    max_num = 10
elif difficulty == "medium":
    max_num = 50
else:
    max_num = 100

score = 0
num_questions = 5

print(f"Starting Quiz Game - Difficulty: {difficulty}")

for i in range(num_questions):
    num1 = random.randint(1, max_num)
    num2 = random.randint(1, max_num)

    print(f"\nQuestion {i+1}: {num1} + {num2} = ?")

    start_time = datetime.datetime.now()

    try:
        answer = int(input("Your answer: "))
    except:
        print("Invalid input!")
        continue

    end_time = datetime.datetime.now()
    time_taken = (end_time - start_time).seconds

    correct_answer = num1 + num2

    if answer == correct_answer:
        # Score based on speed (faster = more score)
        points = max(1, 10 - time_taken)
        score += points
        print(f"Correct! +{points} points")
    else:
        print(f"Wrong! Correct answer was {correct_answer}")

# Final score using math module
final_score = math.ceil(score)
print(f"\nFinal Score: {final_score}")

# Save score using os module
filename = "scores.txt"

if not os.path.exists(filename):
    with open(filename, "w") as f:
        f.write("Scores:\n")

with open(filename, "a") as f:
    f.write(f"Score: {final_score} | Difficulty: {difficulty}\n")

print("Score saved to file!")
```
