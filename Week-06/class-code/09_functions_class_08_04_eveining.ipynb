```python
def factorial(n):

    if n==1:
        return 1
    
    result = n * factorial(n-1)
    return result
```

```python
factorial(4)
```

```python
def countdown(n):
    if n<=0:
        return n
    
    a = countdown(n-1)
    return a + n
```
