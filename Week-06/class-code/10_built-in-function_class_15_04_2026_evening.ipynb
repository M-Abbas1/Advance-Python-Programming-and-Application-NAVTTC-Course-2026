```python
def MAP(function, seq):
    output_list = []

    for el in seq:
        output_list.append(function(el))

    return output_list

    
```

```python
list(map(len, fruits))
```

```python
MAP(len, fruits)
```

```python
def MAP(function, seq):
    return [function(el) for el in seq]
```

```python
MAP(len, fruits)
```

```python
MAP = lambda fun, seq: [fun(el) for el in seq]
```

```python
MAP(len, fruits)
```
