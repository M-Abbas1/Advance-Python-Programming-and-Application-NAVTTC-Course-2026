```python
def countdown(n):
    if n<=0:
        print("Go!")
        return 
    print(n)
    countdown(n-1)

countdown(5)
```

```python
# ── Fibonacci ──────────────────────────────────────────────────
def fibonacci(n):
    """Return the nth Fibonacci number (0-indexed)."""
    if n <= 1:          # base cases
        return n
    return fibonacci(n - 1) + fibonacci(n - 2)

sequence = [fibonacci(i) for i in range(12)]
print("Fibonacci sequence:", sequence)
```
