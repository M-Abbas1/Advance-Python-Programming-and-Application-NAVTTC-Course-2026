```python
student_data_dic = {}

with open("student_data.txt", 'r') as file:
    for line in file:
        list_st_line = line.strip().split(',')
        k = list_st_line[0]

        temp_dic = {}

        temp_dic['name'] = list_st_line[1]
        temp_dic['age'] = int(list_st_line[2])
        temp_dic['dept'] = list_st_line[3]
        temp_dic['gpa'] = float(list_st_line[4])

        student_data_dic[k] = temp_dic

```
