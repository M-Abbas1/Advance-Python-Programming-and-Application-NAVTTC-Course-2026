```python
class MarksError(Exception):
    pass

class MarksLessThanZero(Exception):
    pass

try:
    name = input("Enter student Name : ")
    obtain_marks = int(input("Enter Obtain Marks  "))
    total_marks = int(input("Enter Total Marks : "))


    if obtain_marks > total_marks:
        raise MarksError("Obtain marks can't be greate than total marks")
    
    if total_marks<0 or obtain_marks<0:
        raise MarksLessThanZero("Total Marks are less than zero!")

    percentage = (obtain_marks/total_marks) * 100
except ValueError:
    print("Enter valid input!")
except ZeroDivisionError:
    print("Total Marks are Zero")
except MarksError:
    print("Obtain marks are greater than total marks")
except MarksLessThanZero:
    print("Obtain Marks or toal Marks are less than zero")
else:
    print("------Student Info------")
    print("Student Name : ", name)
    print("Obtained Marks : ", obtain_marks)
    print("Total Marks : ", total_marks)
    print("Percentage : ", percentage)
    print("-"*26)
```
