ASSESSMENT TASK 1: HOTEL ROOM BOOKING SYSTEM
============================================
Create a hotel room booking system that calculates the total cost based on
room type, number of nights, season, and guest preferences.

Requirements:
- Take guest name, room type (Standard/Deluxe/Suite), number of nights
- Ask for season (Summer/Winter/Spring/Fall)
- Ask if breakfast is included (yes/no)
- Calculate total cost with following rates:
  Room Rates:
    * Standard: $100/night
    * Deluxe: $150/night
    * Suite: $250/night
  
  Season Adjustments:
    * Summer (June-Aug): +20% on room rate
    * Winter (Dec-Feb): +15% on room rate
    * Spring/Fall: No adjustment
  
  Additional Services:
    * Breakfast: $15 per night
    * Service charge: 10% of room cost
    * Tax: 8% of total (after service charge)

- Display itemized bill with all calculations
- Show discount if stay is more than 5 nights (10% off room cost)


Expected Output Format: <br>
`========================`  <br>
`--- Hotel Booking System ---`   <br>
`Guest Name: [name]`  <br>
`Room Type (Standard/Deluxe/Suite): [type]`  <br>
`Number of Nights: [nights]`    <br>
`Season (Summer/Winter/Spring/Fall): [season]`   <br>
`Include Breakfast? (yes/no): [choice]`   <br>



`=========== BOOKING RECEIPT ===========` <br>
`Guest: [NAME IN CAPS]`     <br>
`Room Type: [type]`    <br>
`Duration: [nights] nights`    <br>
`-----------------------------------`    <br>
`Room Charges:        $[amount]`   <br>
`Season Adjustment:   $[amount]`   <br>
`Breakfast:           $[amount]`   <br>
`Service Charge:      $[amount]`   <br>
`Tax (8%):            $[amount]`   <br>
`Discount:            $[amount]`   <br>
`-----------------------------------`    <br>
`TOTAL AMOUNT:        $[total]`    <br>
`=======================================`   <br>

NOTES:
- Use proper formatting for currency (2 decimal places)
- Use conditionals to determine room rate
- Use nested conditionals for season adjustment
- Validate input where necessary
- Display appropriate messages

ASSESSMENT TASK 3: SHOPPING CART WITH MEMBERSHIP
================================================

benefits, and promotional offers.

Requirements:
- Take customer name and membership status (Gold/Silver/None)
- Input 3 items with name, quantity, and unit price
- Calculate subtotal
- Apply membership discounts:
  
  Membership Benefits:
    * Gold Member: 20% discount + free shipping
    * Silver Member: 10% discount + $5 shipping
    * No Membership: No discount + $10 shipping
  
  Promotional Offers (applied after membership discount):
    * Cart value > $200: Extra 5% off
    * Cart value > $500: Extra 10% off

- Ask if customer wants gift wrapping
- Display detailed invoice with all calculations

ASSESSMENT TASK 4: EXAM QUALIFICATION SYSTEM
============================================
Create a comprehensive system that determines if a student qualifies for
final exam based on multiple criteria and calculates their current status.

Requirements:
- Take student name, ID, and course name
- Input scores for:
  * Quiz 1 (out of 20)
  * Quiz 2 (out of 20)
  * Assignment 1 (out of 30)
  * Assignment 2 (out of 30)
  * Midterm Exam (out of 100)
- Input attendance percentage
- Calculate and check qualification:
  
  Qualification Rules:
    * Attendance must be >= 75%
    * Average quiz score must be >= 60%
    * Average assignment score must be >= 50%
    * Midterm score must be >= 40%
    * Overall average must be >= 50%
  
  Weights for Overall Average:
    * Quizzes: 15% (combined)
    * Assignments: 25% (combined)
    * Midterm: 60%
  
  Status Categories:
    * Excellent: Overall >= 85% AND all criteria met
    * Good: Overall >= 70% AND all criteria met
    * Satisfactory: Overall >= 60% AND all criteria met
    * At Risk: Overall >= 50% AND all criteria met
    * Not Qualified: Any criterion not met

- Display comprehensive report
- Show which criteria are not met (if any)
- Provide specific recommendations
