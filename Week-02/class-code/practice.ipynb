```python
num = int(input("Enter the nubmer : "))

str_num = str(num)  # 'num'

rev_num_str = str_num[::-1]

rev_num = int(rev_num_str)

print(f"the {num} is palindrome") if num==rev_num else print(f"the {num} is not palindrome")
```

```python
num1 = 1221

str_num1 = str(num1)

l = len(str_num1)

rev_str = str_num1[l-1] + str_num1[l-2] + str_num1[l-3]

rev_num1 = int(rev_str)

print(f"the {num1} is palindrome") if num1==rev_num1 else print(f"the {num1} is not palindrome")
```
