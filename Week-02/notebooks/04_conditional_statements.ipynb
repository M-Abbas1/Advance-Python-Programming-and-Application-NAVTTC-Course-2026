# Conditional Statements

## Introduction

Conditional statements are the building blocks of decision-making in programming. They allow your program to execute different code based on whether certain conditions are true or false. Just like in real life, where you make decisions based on circumstances ("If it's raining, I'll take an umbrella"), conditional statements enable your code to adapt and respond to different situations.

Python provides several ways to implement conditional logic, making your programs dynamic and intelligent!

**What you'll learn:**
- The `if` statement - basic decision making
- The `if-else` statement - choosing between two options
- The `if-elif-else` statement - multiple conditions
- Nested conditional statements - conditions within conditions
- Conditional expressions (ternary operator) - compact conditions
- Common patterns and best practices
- Truthiness and falsiness in Python
- Practical real-world applications

Let's start making decisions in code! 🎯

## Part 1: Comprehensive Practice 

## 1. The if Statement

The simplest form of conditional statement. Code inside the if block executes only if the condition is True.

### 1.1 Basic if Statement

````python
# Basic if statement syntax:
# if condition:
#     code to execute if condition is True

age = 20

if age >= 18:
    print("You are an adult")
    print("You can vote")

print("This line always executes")
````

### 1.2 Indentation is Critical!

````python
# Python uses indentation to define code blocks
# Standard is 4 spaces

temperature = 30

if temperature > 25:
    print("It's hot outside!")  # Indented - part of if block
    print("Stay hydrated")      # Indented - part of if block

print("Have a nice day!")  # Not indented - always executes
````

### 1.3 Multiple Conditions with if

````python
# Using logical operators in conditions
age = 25
has_license = True

if age >= 18 and has_license:
    print("You can drive a car")

# Another example
score = 85

if score >= 60 and score <= 100:
    print("Valid score - You passed!")

# Same using chained comparison
if 60 <= score <= 100:
    print("Valid score (using chained comparison)")
````

### 1.4 Multiple Independent if Statements

````python
# Each if is independent - all can execute
number = 10

if number > 0:
    print("Number is positive")

if number < 20:
    print("Number is less than 20")

if number % 2 == 0:
    print("Number is even")

print("\nAll three conditions were checked!")
````

## 2. The if-else Statement

Provides an alternative path when the condition is False.

### 2.1 Basic if-else

````python
# if-else syntax:
# if condition:
#     code if True
# else:
#     code if False

age = 15

if age >= 18:
    print("You are an adult")
else:
    print("You are a minor")

print("Age check complete")
````

### 2.2 Even or Odd Checker

````python
# Classic example: even or odd
number = 7

if number % 2 == 0:
    print(f"{number} is even")
else:
    print(f"{number} is odd")
````

### 2.3 Multiple Statements in Each Block

````python
temperature = 35

if temperature > 30:
    print("It's very hot!")
    print("Drink plenty of water")
    print("Stay in shade")
else:
    print("Weather is pleasant")
    print("Enjoy your day outside")
````

## 3. The if-elif-else Statement

For handling multiple conditions - only ONE block executes.

### 3.1 Basic if-elif-else

````python
# if-elif-else syntax:
# if condition1:
#     code for condition1
# elif condition2:
#     code for condition2
# else:
#     code if all conditions are False

score = 75

if score >= 90:
    print("Grade: A")
elif score >= 80:
    print("Grade: B")
elif score >= 70:
    print("Grade: C")
elif score >= 60:
    print("Grade: D")
else:
    print("Grade: F")
````

### 3.2 Multiple elif Statements

````python
# You can have as many elif as needed
day = 3

if day == 1:
    print("Monday")
elif day == 2:
    print("Tuesday")
elif day == 3:
    print("Wednesday")
elif day == 4:
    print("Thursday")
elif day == 5:
    print("Friday")
elif day == 6:
    print("Saturday")
elif day == 7:
    print("Sunday")
else:
    print("Invalid day number")
````

### 3.3 Order Matters in if-elif-else

````python
# Python checks conditions in order and stops at first True
number = 15

# Correct order - specific to general
if number > 20:
    print("Greater than 20")
elif number > 10:
    print("Greater than 10 but not more than 20")
elif number > 0:
    print("Greater than 0 but not more than 10")
else:
    print("Zero or negative")

print()

# Wrong order - too general first
if number > 0:
    print("This catches everything positive!")  # This executes
elif number > 10:  # Never reached for positive numbers
    print("This won't execute for 15")
elif number > 20:  # Never reached
    print("This won't execute either")
````

### 3.4 if-elif Without else

````python
# else is optional
age = 30

if age < 13:
    print("Child")
elif age < 20:
    print("Teenager")
elif age < 60:
    print("Adult")
# No else - if age is 60 or above, nothing happens

print("Age category check complete")
````

## 4. Nested Conditional Statements

Conditional statements inside other conditional statements.

### 4.1 Basic Nesting

````python
# Nesting if inside if
age = 25
has_license = True

if age >= 18:
    print("You are old enough")
    
    if has_license:
        print("You can drive!")
    else:
        print("But you need a license")
else:
    print("You are too young to drive")
````

### 4.2 Multiple Levels of Nesting

````python
# Deeper nesting
username = "alice"
password = "secret123"
is_active = True

if username == "alice":
    print("Username correct")
    
    if password == "secret123":
        print("Password correct")
        
        if is_active:
            print("Login successful!")
        else:
            print("Account is inactive")
    else:
        print("Wrong password")
else:
    print("Username not found")
````

### 4.3 Nested if-elif-else

````python
# Complex decision making
age = 25
income = 45000

if age >= 18:
    if income >= 50000:
        print("Eligible for premium credit card")
    elif income >= 30000:
        print("Eligible for standard credit card")
    else:
        print("Eligible for basic credit card")
else:
    print("Not eligible - must be 18 or older")
````

### 4.4 Avoiding Deep Nesting

````python
# Deep nesting can be hard to read
# Better approach: use logical operators

# Instead of deep nesting:
age = 25
has_license = True
has_car = True

# Nested (harder to read)
if age >= 18:
    if has_license:
        if has_car:
            print("You can drive your car")

print()

# Better: use 'and' (easier to read)
if age >= 18 and has_license and has_car:
    print("You can drive your car")
````

## 5. Truthiness and Falsiness

Understanding what Python considers True or False in conditions.

### 5.1 Falsy Values

````python
# Values that evaluate to False:
# - False
# - None
# - 0 (zero)
# - 0.0 (zero float)
# - Empty sequences: '', "", [], ()
# - Empty mappings: {}

# Testing False
if False:
    print("Won't print")
else:
    print("False is falsy")

# Testing None
value = None
if value:
    print("Won't print")
else:
    print("None is falsy")

# Testing zero
if 0:
    print("Won't print")
else:
    print("0 is falsy")

# Testing empty string
if "":
    print("Won't print")
else:
    print("Empty string is falsy")

# Testing empty list
if []:
    print("Won't print")
else:
    print("Empty list is falsy")
````

### 5.2 Truthy Values

````python
# Everything else is truthy!

# Non-zero numbers
if 1:
    print("1 is truthy")

if -5:
    print("-5 is truthy")

if 3.14:
    print("3.14 is truthy")

# Non-empty strings
if "Hello":
    print("Non-empty string is truthy")

if " ":  # Even just a space
    print("String with space is truthy")

# Non-empty collections
if [1, 2, 3]:
    print("Non-empty list is truthy")

if {"key": "value"}:
    print("Non-empty dict is truthy")
````

### 5.3 Practical Use of Truthiness

````python
# Checking if string is not empty
name = "Alice"

if name:  # Better than: if name != ""
    print(f"Hello, {name}!")
else:
    print("No name provided")

# Checking if list has items
items = [1, 2, 3]

if items:  # Better than: if len(items) > 0
    print(f"You have {len(items)} items")
else:
    print("Your list is empty")

# Checking for None
result = None

if result is None:  # Best practice for None check
    print("No result")
else:
    print(f"Result: {result}")
````

## 6. Conditional Expressions (Ternary Operator)

A compact way to write simple if-else statements in one line.

### 6.1 Basic Ternary Operator

````python
# Syntax: value_if_true if condition else value_if_false

# Traditional if-else
age = 20
if age >= 18:
    status = "Adult"
else:
    status = "Minor"
print(f"Status: {status}")

# Same with ternary operator
age = 20
status = "Adult" if age >= 18 else "Minor"
print(f"Status: {status}")
````

### 6.2 More Ternary Examples

````python
# Even or odd
number = 7
result = "Even" if number % 2 == 0 else "Odd"
print(f"{number} is {result}")

# Maximum of two numbers
a = 10
b = 15
maximum = a if a > b else b
print(f"Maximum: {maximum}")

# Minimum of two numbers
minimum = a if a < b else b
print(f"Minimum: {minimum}")

# Absolute value
x = -5
absolute = x if x >= 0 else -x
print(f"Absolute value of {x}: {absolute}")
````

### 6.3 Using Ternary in print()

````python
# Directly in print statements
score = 85
print(f"You {'passed' if score >= 60 else 'failed'} the exam")

# Another example
temperature = 30
print(f"It's {'hot' if temperature > 25 else 'cold'} today")

# With calculations
price = 100
discount = 0.2 if price > 50 else 0.1
final_price = price * (1 - discount)
print(f"Final price: ${final_price}")
````

### 6.4 When NOT to Use Ternary

````python
# Don't use ternary for complex logic - it becomes hard to read

# Bad: too complex
score = 85
grade = "A" if score >= 90 else "B" if score >= 80 else "C" if score >= 70 else "F"
print(f"Grade (hard to read): {grade}")

# Better: use regular if-elif-else for multiple conditions
if score >= 90:
    grade = "A"
elif score >= 80:
    grade = "B"
elif score >= 70:
    grade = "C"
else:
    grade = "F"
print(f"Grade (easy to read): {grade}")
````

## 7. Common Patterns and Best Practices

### 7.1 Guard Clauses

````python
# Guard clauses: check for invalid conditions first

def process_age(age):
    # Guard clauses at the top
    if age < 0:
        print("Error: Age cannot be negative")
        return
    
    if age > 150:
        print("Error: Age seems invalid")
        return
    
    # Main logic
    if age < 18:
        print("Minor")
    else:
        print("Adult")

process_age(25)
process_age(-5)
process_age(200)
````

### 7.2 Avoiding Redundant Conditions

````python
# Bad: redundant condition
x = 10
if x > 5:
    print("x is greater than 5")
else:
    if x <= 5:  # Redundant! If not > 5, must be <= 5
        print("x is 5 or less")

print()

# Good: simple else
if x > 5:
    print("x is greater than 5")
else:  # Implicitly x <= 5
    print("x is 5 or less")
````

### 7.3 Boolean Variables vs Comparisons

````python
is_student = True

# Bad: comparing boolean to True
if is_student == True:
    print("Is a student")

# Good: use boolean directly
if is_student:
    print("Is a student")

# For False check:
# Bad
if is_student == False:
    print("Not a student")

# Good
if not is_student:
    print("Not a student")
````

### 7.4 Simplifying Boolean Returns

````python
# Bad: unnecessary if-else for boolean return
def is_adult_bad(age):
    if age >= 18:
        return True
    else:
        return False

# Good: return the comparison directly
def is_adult_good(age):
    return age >= 18

print(is_adult_good(20))  # True
print(is_adult_good(15))  # False
````

## 8. Practical Real-World Examples

### 8.1 Login System

````python
# Simple login validation
correct_username = "admin"
correct_password = "pass123"

username = "admin"
password = "pass123"

if username == correct_username and password == correct_password:
    print("✓ Login successful!")
    print("Welcome to the system")
elif username != correct_username:
    print("✗ Invalid username")
else:
    print("✗ Invalid password")
````

### 8.2 Discount Calculator

````python
# Calculate discount based on purchase amount
total_amount = 150
is_member = True

# Determine discount percentage
if total_amount >= 200:
    discount = 0.20  # 20% off
elif total_amount >= 100:
    discount = 0.15  # 15% off
elif total_amount >= 50:
    discount = 0.10  # 10% off
else:
    discount = 0.05  # 5% off

# Extra discount for members
if is_member:
    discount += 0.05  # Additional 5% for members

# Calculate final price
discount_amount = total_amount * discount
final_price = total_amount - discount_amount

print(f"Original Amount: ${total_amount}")
print(f"Discount: {discount * 100}%")
print(f"Discount Amount: ${discount_amount}")
print(f"Final Price: ${final_price}")
````

### 8.3 Grade Calculator with Comments

````python
# Comprehensive grade calculator
score = 87
attendance = 92

# Determine letter grade
if score >= 90:
    grade = "A"
    comment = "Excellent!"
elif score >= 80:
    grade = "B"
    comment = "Very Good!"
elif score >= 70:
    grade = "C"
    comment = "Good"
elif score >= 60:
    grade = "D"
    comment = "Needs Improvement"
else:
    grade = "F"
    comment = "Failed"

# Check attendance
if attendance < 75:
    grade += "*"  # Mark for poor attendance
    comment += " (Low Attendance Warning!)"

# Display results
print("=== Grade Report ===")
print(f"Score: {score}")
print(f"Attendance: {attendance}%")
print(f"Grade: {grade}")
print(f"Comment: {comment}")
````

### 8.4 BMI Calculator with Categories

````python
# BMI Calculator with health categories
weight = 70  # kg
height = 1.75  # meters

# Calculate BMI
bmi = weight / (height ** 2)

# Determine category
if bmi < 18.5:
    category = "Underweight"
    advice = "Consider consulting a nutritionist"
elif bmi < 25:
    category = "Normal weight"
    advice = "Maintain your healthy lifestyle!"
elif bmi < 30:
    category = "Overweight"
    advice = "Consider regular exercise"
else:
    category = "Obese"
    advice = "Consult a healthcare professional"

# Display results
print("=== BMI Report ===")
print(f"Weight: {weight} kg")
print(f"Height: {height} m")
print(f"BMI: {bmi:.2f}")
print(f"Category: {category}")
print(f"Advice: {advice}")
````

### 8.5 Ticket Pricing System

````python
# Movie ticket pricing
age = 25
is_student = False
is_weekend = True

# Base price
base_price = 15

# Age-based pricing
if age < 12:
    price = base_price * 0.5  # 50% off for children
    ticket_type = "Child"
elif age >= 65:
    price = base_price * 0.7  # 30% off for seniors
    ticket_type = "Senior"
else:
    price = base_price
    ticket_type = "Adult"

# Student discount
if is_student and age >= 12 and age < 65:
    price *= 0.85  # 15% off for students
    ticket_type += " (Student)"

# Weekend surcharge
if is_weekend:
    price += 3  # $3 weekend surcharge
    ticket_type += " - Weekend"

# Display
print("=== Ticket Information ===")
print(f"Ticket Type: {ticket_type}")
print(f"Price: ${price:.2f}")
````
