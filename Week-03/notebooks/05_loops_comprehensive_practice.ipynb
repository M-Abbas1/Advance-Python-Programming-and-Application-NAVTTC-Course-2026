## 1. The while Loop

The while loop repeats a block of code as long as a condition is True.

### 1.1 Basic while Loop

```python
# Basic while loop syntax:
# while condition:
#     code to repeat

# Print numbers 1 to 5
count = 1

while count <= 5:
    print(f"Count: {count}")
    count = count + 1  # Very important! Update the counter

print("Loop finished!")
```

### 1.2 Understanding the Loop Flow

```python
# Let's see step by step what happens
counter = 1

print("Starting loop...")
while counter <= 3:
    print(f"  Iteration {counter}: Counter is {counter}")
    counter = counter + 1
    print(f"  After increment: Counter is now {counter}")
    print()

print(f"Loop ended. Final counter value: {counter}")
```

### 1.3 Common while Loop Patterns

```python
# Pattern 1: Counting up
print("Counting up:")
i = 1
while i <= 5:
    print(i, end=" ")
    i += 1  # Same as: i = i + 1
print()

print()

# Pattern 2: Counting down
print("Counting down:")
i = 5
while i >= 1:
    print(i, end=" ")
    i -= 1  # Same as: i = i - 1
print()

print()

# Pattern 3: Incrementing by 2
print("Even numbers:")
i = 2
while i <= 10:
    print(i, end=" ")
    i += 2
print()
```

### 1.4 User Input with while Loop

```python
# Keep asking until correct password
correct_password = "python123"
attempts = 0
max_attempts = 3

while attempts < max_attempts:
    password = input(f"Attempt {attempts + 1}/{max_attempts} - Enter password: ")
    
    if password == correct_password:
        print("✓ Access granted!")
        break  # Exit the loop
    else:
        attempts += 1
        remaining = max_attempts - attempts
        if remaining > 0:
            print(f"✗ Wrong password. {remaining} attempts remaining.")
        else:
            print("✗ Account locked. Too many failed attempts.")
```

### 1.5 Accumulator Pattern

```python
# Sum of numbers from 1 to 10
number = 1
total = 0

while number <= 10:
    total = total + number
    print(f"Adding {number}: Total = {total}")
    number += 1

print(f"\nSum of numbers 1 to 10: {total}")
```

### 1.6 Infinite Loops and How to Avoid Them

```python
# WARNING: This would be an infinite loop (DON'T RUN!)
# count = 1
# while count <= 5:
#     print(count)
#     # Missing: count += 1  <- This causes infinite loop!

# To stop infinite loop: Press CTRL+C or stop the kernel

# Correct version:
count = 1
while count <= 5:
    print(count)
    count += 1  # Always update your loop variable!

print("Loop completed safely")
```

## 2. The for Loop

The for loop iterates over a sequence (like a list, string, or range).

### 2.1 Basic for Loop with Lists

```python
# for loop syntax:
# for variable in sequence:
#     code to repeat

# Iterate over a list
fruits = ["apple", "banana", "cherry", "date"]

print("Fruits in my basket:")
for fruit in fruits:
    print(f"- {fruit}")
```

### 2.2 for Loop with Strings

```python
# Strings are sequences of characters
word = "PYTHON"

print("Letters in PYTHON:")
for letter in word:
    print(letter)

print()

# Counting vowels
text = "Hello World"
vowel_count = 0

for char in text.lower():
    if char in "aeiou":
        vowel_count += 1

print(f"Vowels in '{text}': {vowel_count}")
```

### 2.3 The range() Function

```python
# range(stop) - from 0 to stop-1
print("range(5):")
for i in range(5):
    print(i, end=" ")
print()

print()

# range(start, stop) - from start to stop-1
print("range(1, 6):")
for i in range(1, 6):
    print(i, end=" ")
print()

print()

# range(start, stop, step) - with custom increment
print("range(0, 11, 2):")
for i in range(0, 11, 2):
    print(i, end=" ")
print()

print()

# Counting backwards
print("range(10, 0, -1):")
for i in range(10, 0, -1):
    print(i, end=" ")
print()
```

### 2.4 for Loop with range() - Common Patterns

```python
# Print multiplication table of 5
print("Multiplication Table of 5:")
for i in range(1, 11):
    result = 5 * i
    print(f"5 × {i} = {result}")

print()

# Sum of first 10 numbers
total = 0
for num in range(1, 11):
    total += num

print(f"Sum of 1 to 10: {total}")

print()

# Print squares
print("Squares of first 5 numbers:")
for n in range(1, 6):
    square = n ** 2
    print(f"{n}² = {square}")
```

### 2.5 Iterating with Index - enumerate()

```python
# enumerate() gives both index and value
colors = ["Red", "Green", "Blue", "Yellow"]

print("Colors with index:")
for index, color in enumerate(colors):
    print(f"{index}: {color}")

print()

# Start index from 1
print("Colors with index starting from 1:")
for index, color in enumerate(colors, start=1):
    print(f"{index}. {color}")
```

### 2.6 for vs while - When to Use Which?

```python
# Use FOR when:
# - You know how many times to loop
# - You're iterating over a collection

# Example: Print 1 to 5
print("Using for loop:")
for i in range(1, 6):
    print(i, end=" ")
print()

print()

# Use WHILE when:
# - You don't know how many times to loop
# - You loop until a condition becomes False

# Example: Keep doubling until > 100
print("Using while loop:")
number = 1
while number <= 100:
    print(number, end=" ")
    number *= 2
print()
```

## 3. Loop Control Statements

Control the flow of loops with break, continue, and pass.

### 3.1 The break Statement

```python
# break: Exit the loop immediately

# Example 1: Stop when found
numbers = [1, 3, 5, 7, 8, 9, 11]

print("Finding first even number:")
for num in numbers:
    if num % 2 == 0:
        print(f"Found even number: {num}")
        break  # Exit loop
    print(f"  Checking {num}...")

print()

# Example 2: Search in string
text = "Python Programming"
search_char = "g"

for index, char in enumerate(text):
    if char == search_char:
        print(f"Found '{search_char}' at position {index}")
        break
else:
    print(f"'{search_char}' not found")
```

### 3.2 The continue Statement

```python
# continue: Skip rest of current iteration, go to next

# Example 1: Print only odd numbers
print("Odd numbers from 1 to 10:")
for i in range(1, 11):
    if i % 2 == 0:
        continue  # Skip even numbers
    print(i, end=" ")
print()

print()

# Example 2: Skip negative numbers
numbers = [5, -2, 8, -1, 3, -7, 10]

print("Positive numbers only:")
for num in numbers:
    if num < 0:
        continue  # Skip negative
    print(num, end=" ")
print()

print()

# Example 3: Skip vowels
word = "PYTHON"
print(f"Consonants in {word}:")
for letter in word:
    if letter.lower() in "aeiou":
        continue
    print(letter, end=" ")
print()
```

### 3.3 The pass Statement

```python
# pass: Do nothing, placeholder

# Used when syntax requires a statement but you don't want to do anything

for i in range(5):
    if i == 2:
        pass  # Will implement later
    else:
        print(i)

print()

# More practical example
numbers = [1, 2, 3, 4, 5]

for num in numbers:
    if num % 2 == 0:
        pass  # Even numbers - will process later
    else:
        print(f"{num} is odd")
```

### 3.4 The else Clause with Loops

```python
# else with for/while: Executes if loop completes normally (no break)

# Example 1: Search with for-else
numbers = [1, 3, 5, 7, 9]
search = 6

for num in numbers:
    if num == search:
        print(f"Found {search}!")
        break
else:
    print(f"{search} not found in list")

print()

# Example 2: while-else
password = "secret"
attempts = 0

while attempts < 3:
    user_input = input(f"Attempt {attempts + 1}: Enter password: ")
    if user_input == password:
        print("Access granted!")
        break
    attempts += 1
else:
    print("Too many failed attempts. Account locked.")
```

## 4. Nested Loops

Loops inside other loops.

### 4.1 Basic Nested Loops

```python
# Outer loop runs, for each iteration inner loop runs completely

print("Nested loop example:")
for i in range(1, 4):
    print(f"Outer loop: i = {i}")
    for j in range(1, 3):
        print(f"  Inner loop: j = {j}")
    print()
```

### 4.2 Multiplication Table

```python
# Print multiplication table 1-5
print("Multiplication Table (1-5):")
print("=" * 30)

for i in range(1, 6):
    for j in range(1, 6):
        result = i * j
        print(f"{result:3}", end=" ")  # :3 means width of 3
    print()  # New line after inner loop
```

### 4.3 Pattern Printing

```python
# Pattern 1: Right triangle
print("Pattern 1:")
for i in range(1, 6):
    for j in range(i):
        print("*", end="")
    print()

print()

# Pattern 2: Number triangle
print("Pattern 2:")
for i in range(1, 6):
    for j in range(1, i + 1):
        print(j, end="")
    print()

print()

# Pattern 3: Square
print("Pattern 3:")
for i in range(4):
    for j in range(5):
        print("#", end=" ")
    print()
```

### 4.4 Nested Loops with Lists

```python
# List of lists
matrix = [
    [1, 2, 3],
    [4, 5, 6],
    [7, 8, 9]
]

print("Matrix:")
for row in matrix:
    for element in row:
        print(element, end=" ")
    print()

print()

# Calculate sum of all elements
total = 0
for row in matrix:
    for element in row:
        total += element

print(f"Sum of all elements: {total}")
```

## 5. Practical Applications

### 5.1 Menu-Driven Program

```python
# Simple calculator with menu
print("=== Simple Calculator ===")

running = True
while running:
    print("\n1. Add")
    print("2. Subtract")
    print("3. Multiply")
    print("4. Divide")
    print("5. Exit")
    
    choice = input("Enter choice (1-5): ")
    
    if choice == "5":
        print("Thank you! Goodbye.")
        running = False
    elif choice in ["1", "2", "3", "4"]:
        num1 = float(input("Enter first number: "))
        num2 = float(input("Enter second number: "))
        
        if choice == "1":
            result = num1 + num2
            print(f"Result: {num1} + {num2} = {result}")
        elif choice == "2":
            result = num1 - num2
            print(f"Result: {num1} - {num2} = {result}")
        elif choice == "3":
            result = num1 * num2
            print(f"Result: {num1} × {num2} = {result}")
        elif choice == "4":
            if num2 != 0:
                result = num1 / num2
                print(f"Result: {num1} ÷ {num2} = {result:.2f}")
            else:
                print("Error: Division by zero!")
    else:
        print("Invalid choice! Please try again.")
```

### 5.2 Number Guessing Game

```python
# Guess the number game
import random

print("=== Number Guessing Game ===")
print("I'm thinking of a number between 1 and 100")

secret_number = random.randint(1, 100)
attempts = 0
max_attempts = 7

while attempts < max_attempts:
    guess = int(input(f"\nAttempt {attempts + 1}/{max_attempts} - Enter your guess: "))
    attempts += 1
    
    if guess == secret_number:
        print(f"🎉 Congratulations! You guessed it in {attempts} attempts!")
        break
    elif guess < secret_number:
        print("📈 Too low! Try a higher number.")
    else:
        print("📉 Too high! Try a lower number.")
else:
    print(f"\n😞 Game Over! The number was {secret_number}")
```

### 5.3 Grade Statistics Calculator

```python
# Calculate statistics for student grades
print("=== Grade Statistics Calculator ===")

num_students = int(input("How many students? "))

total = 0
highest = 0
lowest = 100
passed = 0
failed = 0

for i in range(1, num_students + 1):
    grade = float(input(f"Enter grade for student {i}: "))
    
    # Add to total
    total += grade
    
    # Track highest and lowest
    if grade > highest:
        highest = grade
    if grade < lowest:
        lowest = grade
    
    # Count pass/fail
    if grade >= 60:
        passed += 1
    else:
        failed += 1

# Calculate average
average = total / num_students

# Display results
print("\n" + "=" * 35)
print("       STATISTICS REPORT")
print("=" * 35)
print(f"Total Students:    {num_students}")
print(f"Average Grade:     {average:.2f}")
print(f"Highest Grade:     {highest}")
print(f"Lowest Grade:      {lowest}")
print(f"Passed:            {passed} ({passed/num_students*100:.1f}%)")
print(f"Failed:            {failed} ({failed/num_students*100:.1f}%)")
print("=" * 35)
```

### 5.4 Shopping List Manager

```python
# Simple shopping list
print("=== Shopping List Manager ===")

shopping_list = []

while True:
    print("\n1. Add item")
    print("2. View list")
    print("3. Remove item")
    print("4. Clear list")
    print("5. Exit")
    
    choice = input("Choose option: ")
    
    if choice == "1":
        item = input("Enter item name: ")
        shopping_list.append(item)
        print(f"✓ '{item}' added to list")
    
    elif choice == "2":
        if shopping_list:
            print("\nYour Shopping List:")
            for index, item in enumerate(shopping_list, start=1):
                print(f"{index}. {item}")
        else:
            print("List is empty!")
    
    elif choice == "3":
        if shopping_list:
            print("\nCurrent list:")
            for index, item in enumerate(shopping_list, start=1):
                print(f"{index}. {item}")
            pos = int(input("Enter item number to remove: "))
            if 1 <= pos <= len(shopping_list):
                removed = shopping_list.pop(pos - 1)
                print(f"✓ '{removed}' removed")
            else:
                print("Invalid position!")
        else:
            print("List is empty!")
    
    elif choice == "4":
        shopping_list.clear()
        print("✓ List cleared")
    
    elif choice == "5":
        print("Thank you! Happy shopping!")
        break
    
    else:
        print("Invalid choice!")
```

### 5.5 Factorial Calculator

```python
# Calculate factorial using loop
print("=== Factorial Calculator ===")

number = int(input("Enter a number: "))

if number < 0:
    print("Factorial is not defined for negative numbers")
elif number == 0 or number == 1:
    print(f"{number}! = 1")
else:
    factorial = 1
    calculation = ""
    
    for i in range(1, number + 1):
        factorial *= i
        if i == 1:
            calculation = str(i)
        else:
            calculation += f" × {i}"
    
    print(f"\n{number}! = {calculation}")
    print(f"{number}! = {factorial}")
```
