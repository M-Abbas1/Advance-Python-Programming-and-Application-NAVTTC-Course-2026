```python
num_list = [34, 53, 74, 13, 75, 12,  86]
num_we_want = 13

for ind, val in enumerate(num_list):
    print("count")
    if val == num_we_want:
        print(f"Val : {val}")
        print(f"Index : {ind}")
        break
```

```python
p = "programming"

run_counter = 0

for ind, char in enumerate(p):
    run_counter += 1
    print(f"Execution {run_counter}")
    if char == 'g':
        print("char : ", char)
        print("index : ", ind)
        break
```
