## Important Tips

### 1. List Copying
```python
# WRONG - both variables point to same list
list2 = list1
list2.append(5)  # Also changes list1!

# CORRECT - creates actual copy
list2 = list1.copy()
list2 = list(list1)
list2 = list1[:]
```

### 2. Parallel Lists
```python
# Always update all lists together
names = ["Alice", "Bob"]
ages = [25, 30]

# When removing Bob, remove from both
index = names.index("Bob")
names.pop(index)
ages.pop(index)  # Must remove from both!
```

### 3. Don't Modify While Iterating
```python
# WRONG
for item in my_list:
    if condition:
        my_list.remove(item)  # Can skip elements!

# CORRECT - iterate over copy
for item in my_list.copy():
    if condition:
        my_list.remove(item)
```

### 4. Common List Methods
- **Add:** `append()`, `insert()`, `extend()`
- **Remove:** `remove()`, `pop()`, `del`, `clear()`
- **Search:** `in`, `index()`, `count()`
- **Sort:** `sort()`, `sorted()`, `reverse()`

### 5. Testing Checklist
- Empty list: `[]`
- Single element: `[1]`
- Duplicates: `[1, 1, 2, 2]`
- Negative numbers: `[-5, -2, 3]`
