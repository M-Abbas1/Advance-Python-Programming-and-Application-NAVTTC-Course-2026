### Calculator With History

```python
print("=----=Calculator With History=----=")
print("1. Add")
print("2. Subtract")
print("3. Multiply")
print("4. Divide")
print("5. Show Hisotry")
print("6. Clear History")
print("7. Exit")

op = input("\nChoose an Option : ")

if op == '1':
    print("add")
elif op == '2':
    print("subtract")
elif op == '3':
    print("Multiply")
elif op == '4':
    print('Divide')
elif op == '5':
    print("show history")
elif op == '6':
    print("Clear Hisotry")
elif op == '7':
    print("Exit")
else:
    print("Invalid option, try again")
```
