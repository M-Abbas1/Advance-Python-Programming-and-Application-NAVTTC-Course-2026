```python
for i in range(1, 5):        # i = 1+1,  j = 1 --- 1+1  diff  1
    for j in range(1, i+1):  # i = 2+1,  j = 1 --- 2+1  diff  2
        print(j, end=' ')    # i = 3+1,  j = 1 --- 3+1  diff  3
    print()                  # i = 4+1,  j = 1 --- 4+1  diff  4
```
