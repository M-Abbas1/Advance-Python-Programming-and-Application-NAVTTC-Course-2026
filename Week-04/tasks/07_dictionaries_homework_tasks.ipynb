## Important Tips

### 1. Safe Key Access
```python
# Risky - causes KeyError if key missing
value = my_dict["key"]

# Safe - returns None or default
value = my_dict.get("key", default_value)
```

### 2. Check Before Access
```python
if "key" in my_dict:
    value = my_dict["key"]
```

### 3. Iterating
```python
# Best for key-value pairs
for key, value in my_dict.items():
    print(f"{key}: {value}")
```

### 4. Counting Pattern
```python
counts = {}
for item in items:
    counts[item] = counts.get(item, 0) + 1
```

### 5. Nested Access
```python
# Safe nested access
if "outer" in my_dict:
    if "inner" in my_dict["outer"]:
        value = my_dict["outer"]["inner"]
```

### Common Methods
- `get(key, default)` - Safe access
- `update(other_dict)` - Merge dicts
- `pop(key)` - Remove and return
- `keys()`, `values()`, `items()` - Iteration
- `setdefault(key, default)` - Get or set
