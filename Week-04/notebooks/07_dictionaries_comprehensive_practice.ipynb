# Dictionaries in Python

## Introduction

Dictionaries are powerful data structures that store data in key-value pairs. Think of a real dictionary where you look up a word (key) to find its definition (value). In Python, dictionaries let you quickly find values by their unique keys, making them perfect for storing related information like student records, product catalogs, or configuration settings.

Unlike lists that use numeric indices, dictionaries use meaningful keys. Instead of remembering that student grades are at index 3, you can simply use `grades['Alice']` - much more readable and intuitive!

**What makes dictionaries special:**
- **Key-Value Pairs** - Each item has a key and associated value
- **Fast Lookup** - Find values instantly by their key
- **Mutable** - Add, modify, or remove items
- **Flexible** - Keys and values can be different data types
- **Unordered** - Items don't have a fixed position (Python 3.7+ maintains insertion order)

## 1. Creating Dictionaries

Different ways to create dictionaries in Python.

### 1.1 Basic Dictionary Creation

```python
# Empty dictionary
empty_dict = {}
print("Empty:", empty_dict)

# Dictionary with string keys
student = {
    "name": "Alice",
    "age": 20,
    "grade": "A"
}
print("Student:", student)

# Dictionary with different key types
mixed = {
    "name": "Bob",
    1: "first",
    2: "second",
    (1, 2): "tuple key"
}
print("Mixed keys:", mixed)

# Dictionary with various value types
person = {
    "name": "John",
    "age": 25,
    "height": 5.9,
    "married": False,
    "hobbies": ["reading", "coding"],
    "address": {"city": "New York", "zip": "10001"}
}
print("Person:", person)
```

### 1.2 Creating Dictionaries with dict() Constructor

```python
# Using dict() with keyword arguments
student = dict(name="Alice", age=20, grade="A")
print("Using dict():", student)

# From list of tuples
pairs = [("apple", 5), ("banana", 3), ("orange", 7)]
fruits = dict(pairs)
print("From tuples:", fruits)

# From two lists using zip()
keys = ["name", "age", "city"]
values = ["Bob", 22, "Boston"]
person = dict(zip(keys, values))
print("Using zip():", person)
```

## 2. Accessing Dictionary Elements

Different ways to access values in a dictionary.

### 2.1 Accessing Values by Key

```python
student = {
    "name": "Alice",
    "age": 20,
    "grade": "A",
    "gpa": 3.8
}

# Using square brackets
print("Name:", student["name"])
print("Age:", student["age"])

# Using square brackets with non-existent key causes error
# print(student["major"])  # KeyError!

print()

# Using get() method (safe)
print("Name:", student.get("name"))
print("Major:", student.get("major"))  # Returns None, no error
print("Major:", student.get("major", "Not specified"))  # Default value

print()

# Checking if key exists
if "name" in student:
    print("Name exists:", student["name"])

if "major" not in student:
    print("Major not found")
```

### 2.2 Getting All Keys, Values, and Items

```python
student = {"name": "Alice", "age": 20, "grade": "A"}

# Get all keys
keys = student.keys()
print("Keys:", keys)
print("Keys as list:", list(keys))

print()

# Get all values
values = student.values()
print("Values:", values)
print("Values as list:", list(values))

print()

# Get all items (key-value pairs)
items = student.items()
print("Items:", items)
print("Items as list:", list(items))
```

## 3. Modifying Dictionaries

Adding, updating, and removing dictionary items.

### 3.1 Adding and Updating Items

```python
# Adding new items
student = {"name": "Alice", "age": 20}
print("Original:", student)

student["grade"] = "A"
print("After adding grade:", student)

student["gpa"] = 3.8
print("After adding gpa:", student)

print()

# Updating existing items
student["age"] = 21
print("After updating age:", student)

student["grade"] = "A+"
print("After updating grade:", student)
```

### 3.2 Using update() Method

```python
# Update multiple items at once
student = {"name": "Alice", "age": 20}
print("Original:", student)

# Update with another dictionary
student.update({"grade": "A", "gpa": 3.8})
print("After update:", student)

# Update and add new items
student.update({"age": 21, "major": "CS"})
print("After another update:", student)

print()

# Update with keyword arguments
student.update(city="Boston", year=3)
print("After keyword update:", student)
```

### 3.3 Removing Items

```python
student = {
    "name": "Alice",
    "age": 20,
    "grade": "A",
    "gpa": 3.8,
    "city": "Boston"
}

# pop() - remove and return value
print("Original:", student)
grade = student.pop("grade")
print("Removed grade:", grade)
print("After pop:", student)

print()

# pop() with default (if key doesn't exist)
major = student.pop("major", "Not specified")
print("Major:", major)

print()

# del - delete specific key
del student["city"]
print("After del:", student)

print()

# popitem() - remove and return last item
item = student.popitem()
print("Removed item:", item)
print("After popitem:", student)

print()

# clear() - remove all items
student.clear()
print("After clear:", student)
```

## 4. Iterating Through Dictionaries

### 4.1 Iterating Over Keys

```python
student = {"name": "Alice", "age": 20, "grade": "A"}

# Default iteration is over keys
print("Keys:")
for key in student:
    print(f"  {key}")

print()

# Explicit keys() method
print("Using keys():")
for key in student.keys():
    print(f"  {key}: {student[key]}")
```

### 4.2 Iterating Over Values

```python
student = {"name": "Alice", "age": 20, "grade": "A"}

print("Values:")
for value in student.values():
    print(f"  {value}")
```

### 4.3 Iterating Over Key-Value Pairs

```python
student = {"name": "Alice", "age": 20, "grade": "A", "gpa": 3.8}

# Using items() method (most common)
print("Student Information:")
for key, value in student.items():
    print(f"  {key}: {value}")

print()

# Formatted display
print("=== Student Record ===")
for key, value in student.items():
    print(f"{key.capitalize():<10}: {value}")
```

## 5. Dictionary Methods

### 5.1 Copying Dictionaries

```python
# Assignment doesn't create a copy
dict1 = {"a": 1, "b": 2}
dict2 = dict1  # Both point to same dictionary

dict2["c"] = 3
print("Dict1:", dict1)  # Changed!
print("Dict2:", dict2)

print()

# Method 1: copy()
dict1 = {"a": 1, "b": 2}
dict2 = dict1.copy()

dict2["c"] = 3
print("Dict1:", dict1)  # Unchanged
print("Dict2:", dict2)

print()

# Method 2: dict() constructor
dict1 = {"a": 1, "b": 2}
dict2 = dict(dict1)

dict2["c"] = 3
print("Dict1:", dict1)  # Unchanged
print("Dict2:", dict2)
```

### 5.2 setdefault() Method

```python
# setdefault() - get value or set default if key doesn't exist
student = {"name": "Alice", "age": 20}

# Key exists - returns value
name = student.setdefault("name", "Unknown")
print("Name:", name)
print("Student:", student)

print()

# Key doesn't exist - sets and returns default
grade = student.setdefault("grade", "Not graded")
print("Grade:", grade)
print("Student:", student)  # 'grade' was added!
```

### 5.3 fromkeys() Method

```python
# Create dictionary with same value for all keys
keys = ["a", "b", "c"]
dict1 = dict.fromkeys(keys, 0)
print("All zeros:", dict1)

# Default value is None
dict2 = dict.fromkeys(keys)
print("All None:", dict2)

# Practical example
students = ["Alice", "Bob", "Charlie"]
grades = dict.fromkeys(students, "Not graded")
print("Initial grades:", grades)
```

## 6. Dictionary Comprehensions

### 6.1 Basic Dictionary Comprehension

```python
# Create squares dictionary
squares = {x: x**2 for x in range(1, 6)}
print("Squares:", squares)

# Convert list to dictionary with indices
fruits = ["apple", "banana", "cherry"]
fruit_dict = {i: fruit for i, fruit in enumerate(fruits)}
print("Fruits:", fruit_dict)

# Create from two lists
names = ["Alice", "Bob", "Charlie"]
ages = [20, 22, 21]
people = {name: age for name, age in zip(names, ages)}
print("People:", people)
```

### 6.2 Dictionary Comprehension with Conditions

```python
# Only even numbers
even_squares = {x: x**2 for x in range(1, 11) if x % 2 == 0}
print("Even squares:", even_squares)

# Filter dictionary
scores = {"Alice": 85, "Bob": 72, "Charlie": 90, "Diana": 65}
passed = {name: score for name, score in scores.items() if score >= 70}
print("Passed:", passed)

# With if-else
numbers = {x: ("even" if x % 2 == 0 else "odd") for x in range(1, 6)}
print("Even/Odd:", numbers)
```

## 7. Nested Dictionaries

### 7.1 Creating and Accessing Nested Dictionaries

```python
# Nested dictionary example
students = {
    "student1": {
        "name": "Alice",
        "age": 20,
        "grades": {"math": 90, "english": 85}
    },
    "student2": {
        "name": "Bob",
        "age": 22,
        "grades": {"math": 88, "english": 92}
    }
}

# Accessing nested values
print("Student 1 name:", students["student1"]["name"])
print("Student 1 math grade:", students["student1"]["grades"]["math"])
print("Student 2 english:", students["student2"]["grades"]["english"])
```

### 7.2 Iterating Through Nested Dictionaries

```python
students = {
    "Alice": {"age": 20, "grade": "A"},
    "Bob": {"age": 22, "grade": "B"},
    "Charlie": {"age": 21, "grade": "A"}
}

print("=== Student Records ===")
for name, info in students.items():
    print(f"\n{name}:")
    for key, value in info.items():
        print(f"  {key}: {value}")
```

## 8. Common Dictionary Patterns

### 8.1 Counting with Dictionaries

```python
# Count letter frequency
text = "hello world"
frequency = {}

for char in text:
    if char != " ":
        frequency[char] = frequency.get(char, 0) + 1

print("Letter frequency:", frequency)

print()

# Count word occurrences
words = ["apple", "banana", "apple", "cherry", "banana", "apple"]
word_count = {}

for word in words:
    word_count[word] = word_count.get(word, 0) + 1

print("Word count:", word_count)
```

### 8.2 Grouping with Dictionaries

```python
# Group students by grade
students = [
    {"name": "Alice", "grade": "A"},
    {"name": "Bob", "grade": "B"},
    {"name": "Charlie", "grade": "A"},
    {"name": "Diana", "grade": "B"}
]

by_grade = {}
for student in students:
    grade = student["grade"]
    if grade not in by_grade:
        by_grade[grade] = []
    by_grade[grade].append(student["name"])

print("Students by grade:")
for grade, names in by_grade.items():
    print(f"  Grade {grade}: {names}")
```

### 8.3 Swapping Keys and Values

```python
# Original dictionary
original = {"a": 1, "b": 2, "c": 3}
print("Original:", original)

# Swap keys and values
swapped = {value: key for key, value in original.items()}
print("Swapped:", swapped)
```

## 9. Practical Applications

### 9.1 Contact Book

```python
# Simple contact book
contacts = {
    "Alice": {"phone": "555-1234", "email": "alice@email.com"},
    "Bob": {"phone": "555-5678", "email": "bob@email.com"},
    "Charlie": {"phone": "555-9012", "email": "charlie@email.com"}
}

# Display all contacts
print("=== Contact Book ===")
for name, info in contacts.items():
    print(f"\n{name}:")
    print(f"  Phone: {info['phone']}")
    print(f"  Email: {info['email']}")

# Search contact
search = "Bob"
if search in contacts:
    print(f"\nFound {search}:")
    print(f"  Phone: {contacts[search]['phone']}")
```

### 9.2 Product Inventory

```python
# Product inventory system
inventory = {
    "laptop": {"price": 999.99, "quantity": 15},
    "mouse": {"price": 25.50, "quantity": 50},
    "keyboard": {"price": 79.99, "quantity": 30}
}

# Display inventory
print("=== Inventory Report ===")
total_value = 0

for product, details in inventory.items():
    value = details["price"] * details["quantity"]
    total_value += value
    print(f"{product.capitalize()}:")
    print(f"  Price: ${details['price']:.2f}")
    print(f"  Quantity: {details['quantity']}")
    print(f"  Total Value: ${value:.2f}")
    print()

print(f"Total Inventory Value: ${total_value:.2f}")
```

### 9.3 Grade Book System

```python
# Student grade book
gradebook = {
    "Alice": [85, 90, 88, 92],
    "Bob": [78, 85, 80, 88],
    "Charlie": [92, 95, 90, 94]
}

print("=== Grade Report ===")
for student, grades in gradebook.items():
    average = sum(grades) / len(grades)
    highest = max(grades)
    lowest = min(grades)
    
    print(f"\n{student}:")
    print(f"  Grades: {grades}")
    print(f"  Average: {average:.2f}")
    print(f"  Highest: {highest}")
    print(f"  Lowest: {lowest}")
```
