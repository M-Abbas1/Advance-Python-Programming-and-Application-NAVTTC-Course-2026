```python
class Person:
    def __init__(self, name, age, email):
        self.name = name
        self.age = age
        self.email = email

    def get_role(self):
        pass

    def get_summary(self):
        pass

    def __str__(self):
        return f"Name : {self.name}, Age : {self.age}, Email : {self.email}"
```

```python
class Student:
    def __init__(self, student_id, cls):
        self.student_id = student_id
        self.cls = cls
        self.__grades = {}

    def get_role(self):
        return "Student"
    
    def add_grade(self, subject, grade):
        self.__grades[subject] = grade
        print('grades added')

    def show_grades(self):
        print("-------Student Reports-------")
        print("Student ID : ", self.student_id)
        print("Student Class : ", self.cls)
        
        print("-"*30)
        for sub, grd in self.__grades.items():
            print(sub, "  :  ", grd)
        
        print("-"*30)



```

```python
stu1 = Student(1, 8)
```

```python
print(stu1.student_id)
print(stu1.cls)
```

```python
stu1.get_role()
```

```python
stu1.show_grades()
```

```python
stu1.add_grade('python', 90)
stu1.add_grade('java', 90)
stu1.add_grade('C++', 90)
stu1.add_grade('Go', 90)
stu1.add_grade('Cobel', 90)
```

```python
stu1.show_grades()
```

```python
class Person:
    def __init__(self, name, age, email):
        self.name = name
        self.age = age
        self.email = email

    def get_role(self):
        print("get_role function in Person class")
        
    def get_summary(self):
        print("We are in Parent Person class")

    def __str__(self):
        return f"Name : {self.name}, Age : {self.age}, Email : {self.email}"
```

```python
class Student(Person):
    def __init__(self, student_id, name, age, cls, email):
        super().__init__(name, age, email)
        self.student_id = student_id
        self.cls = cls
        self.__grades = {}

    def get_role(self):
        return "Student"
    
    def add_grade(self, subject, grade):
        self.__grades[subject] = grade
        print('grades added')

    def show_grades(self):
        print("-------Student Reports-------")
        print("Student ID : ", self.student_id)
        print("Student Class : ", self.cls)
        
        print("-"*30)
        for sub, grd in self.__grades.items():
            print(sub, "  :  ", grd)
        
        print("-"*30)



```

```python
stu101 = Student(1, 'ali', 21, 9, '123@gmail.com')
```

```python
stu101.get_summary()
```

```python
stu101.get_role()
```

```python
print(stu101)
```

```python
stu101.show_grades()
```

```python
stu101.add_grade('python', 90)
stu101.add_grade('java', 90)
stu101.add_grade('C++', 90)
stu101.add_grade('Go', 90)
stu101.add_grade('Cobel', 90)
```
