```python
class LIST(list):
    def __add__(self, other):
        A = self
        B = other 
        
        if len(A) > len(B):
            A, B = B, A
            
        result = []
        for ind in range(len(A)):
            result.append(A[ind] + B[ind])

        result.extend(B[ind+1:])

        return result
    
    def __mul__(self, other):
        A = self
        B = other 
        
        if len(A) > len(B):
            A, B = B, A
            
        result = []
        for ind in range(len(A)):
            result.append(A[ind] * B[ind])

        result.extend(B[ind+1:])

        return result
```

## Multiple Inheritance

```python
class Clock:
    def __init__(self, hours, minutes, seconds):
        self.hours = hours
        self.minutes = minutes
        self.seconds = seconds

    def display_time(self):
        print(f"{self.hours}:{self.minutes}:{self.seconds}")


class Calendar:
    def __init__(self, year, month, day):
        self.year = year
        self.month = month
        self.day = day

    def show_date(self):
        print(f"{self.day}-{self.month}-{self.year}")
```

```python
c = Clock(4, 50, 30)
```

```python
c.display_time()
```

```python
calendar = Calendar(2026, 4, 24)
```

```python
calendar.show_date()
```

```python
class Clock:
    def __init__(self, hours, minutes, seconds):
        self.hours = hours
        self.minutes = minutes
        self.seconds = seconds

    def display_time(self):
        print(f"{self.hours}:{self.minutes}:{self.seconds}")


class Calendar:
    def __init__(self, year, month, day):
        self.year = year
        self.month = month
        self.day = day

    def show_date(self):
        print(f"{self.day}-{self.month}-{self.year}")


class ClockCalendar(Clock, Calendar):
    def __init__(self, hours, minutes, seconds, year, month, day):
        Clock.__init__(self, hours, minutes, seconds)   
        Calendar.__init__(self, year, month, day)  


```

```python
cc = ClockCalendar(5, 0, 0, 2026, 4, 24)
```

```python
print(cc.hours)
print(cc.minutes)
print(cc.seconds)
print(cc.year)
print(cc.month)
print(cc.day)
```

```python
cc.display_time()
```

```python
cc.show_date()
```
